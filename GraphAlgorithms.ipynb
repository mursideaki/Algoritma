# **DFS**

DFS (Derin Öncelikli Arama), graflarda ve ağaçlarda derinlemesine arama yapmak için kullanılan temel algoritmalardan biridir. Arama algoritmalarının amacı, bir uzay veya yapı içerisinde belirli bir öğeyi bulmaktır ve bu aramalar genellikle en verimli ve hızlı şekilde yapılmaya çalışılır. DFS, bu amaç doğrultusunda derinlemesine bir yaklaşım sergiler ve düğümleri (node) olabildiğince derine giderek araştırır.

**DFS Nasıl Çalışır?**


DFS, graf veya ağaç üzerinde bir başlangıç düğümünden başlar ve mümkün olan her bir düğümün en derinine kadar gider. Bir düğümün tüm komşuları ziyaret edilene kadar derine iner ve geri adım atmadan, tüm düğümler bitene kadar aynı işlemi sürdürür. Herhangi bir düğüm yalnızca bir kez ziyaret edilir.


Örnek:

Elimizde bir graf olduğunu düşünelim. Başlangıç noktamız A düğümü. A'dan başlayarak, komşularına bakarız ve herhangi bir komşu düğüme gideriz. Bu noktada en derine inme prensibinden ötürü DFS, ilk gördüğü komşuya ilerler ve tüm komşuları ziyaret edene kadar devam eder. Örneğin, A -> B -> C yolunu takip edersek, C düğümünden de bir komşuya gidilene kadar aramaya devam ederiz. Bir düğümün komşusu kalmadığında, bir önceki düğüme dönerek diğer komşuları inceleriz.

DFS’in önemli bir parçası, düğümlerin yalnızca bir kez ziyaret edilmesidir. Bu, algoritmanın sonsuz döngülerde kalmasını engeller ve gezilen düğümlerin takip edilmesini sağlar. Düğümler genellikle bir "visited" listesi veya işaretleme yöntemi ile takip edilir.

**Veri Yapısı ve Algoritmanın İşleyişi**

DFS algoritması genellikle bir yığın (stack) veri yapısı kullanılarak uygulanır. Yığın, son giren ilk çıkar (LIFO – Last In First Out) prensibine dayanan bir veri yapısıdır. DFS sırasında bir düğüm yığına eklenir (push) ve geri adım atıldığında yığından çıkarılır (pop). Bu, algoritmanın derinlemesine gitmesini sağlar. Yığın kullanılmadan da DFS uygulanabilir, ancak yığın mantığı DFS'nin temel mekanizmasını daha anlaşılır hale getirir.


DFS'te Yığın Kullanımının Önemi
DFS (Derinlik Öncelikli Arama) algoritması, bir graf veya ağaç yapısında derinlere doğru ilerleyerek tüm düğümleri ziyaret eden bir arama algoritmasıdır. Bu algoritmanın temel veri yapısı olarak yığın (stack) tercih edilir. Peki, neden yığın? Gelin, bu sorunun cevabını detaylı bir şekilde inceleyelim.

##Yığın ve DFS Arasındaki İlişki

###LIFO Özelliği

Yığın, son giren ilk çıkar (LIFO) prensibine dayanır. Bu, en son eklenen elemanın ilk olarak çıkarıldığı anlamına gelir. DFS algoritması da benzer bir mantıkla çalışır. Bir düğüme ulaştığında, önce o düğümün komşularını keşfetmeye çalışır. Yani, daha önce ziyaret edilmemiş bir komşu bulduğunda, bu komşuyu yığına ekler ve onun alt dallarını keşfetmeye başlar.

###Derinlemesine Arama

DFS, bir dalın en derin noktasına kadar inmek ister. Yığın, bu derinlemesine inme işlemini destekler. Bir düğümün tüm alt dalları incelenene kadar, o düğüm yığında kalır. Böylece, algoritma bir dalı tamamen keşfettikten sonra geri döner ve başka bir dalı inceler.

###Geri Dönme Mekanizması

DFS, bir dalın sonuna geldiğinde geri dönmek zorundadır. Yığın, bu geri dönüş işlemini kolaylaştırır. Yığının tepesindeki düğüm, şu anda incelenen dalın en son ziyaret edilen düğümüdür. Bu düğümün tüm komşuları ziyaret edildiğinde, yığından çıkarılır ve bir önceki düğüme dönülür.

##Yığın Kullanmanın Avantajları

Derinlemesine Arama: Yığın, DFS'in temel özelliği olan derinlemesine aramayı doğal olarak destekler.

Geri Dönüş: Yığın, bir dalın sonuna gelindiğinde geri dönüş işlemini kolaylaştırır.

Hafıza Kullanımı: DFS'te genellikle yığın, ziyaret edilen düğümleri tutmak için kullanılır. Bu, hafıza kullanımını optimize eder.

Sonuç olarak Yığın, DFS algoritması için ideal bir veri yapısıdır. LIFO özelliği sayesinde, DFS'in derinlemesine arama özelliği kolayca gerçekleştirilir. Ayrıca, yığın, geri dönüş işlemini ve ziyaret edilen düğümlerin takibini de kolaylaştırır. Bu sayede, DFS algoritması, graf ve ağaç yapılarında birçok farklı problem için etkili bir çözüm sunar.

###Özetle:

Yığın, geri dönüş işlemini kolaylaştırır.

Yığın, ziyaret edilen düğümleri takip etmek için kullanılır.

DFS algoritmasında yığın kullanmanın önemi, algoritmanın temel çalışma prensibi ile doğrudan ilişkilidir. Yığın, DFS'in verimli ve etkili bir şekilde çalışmasını sağlar.


##**Algoritmanın çalışma şekli**

Bir başlangıç düğümü seçilir (root node).

Bu düğümün komşularına gidilir ve ilk komşu düğüm işleme alınır.

Her bir komşu düğüm tekrar ziyaret edilmeyecek şekilde işaretlenir.

Ziyaret edilen düğümün komşularına bakılır, bu adım tüm düğümler bitene kadar devam eder.

Bir düğümün komşuları bittiğinde, önceki düğüme geri dönülür ve bu döngü, graf tamamen gezilene kadar sürer.

##**DFS Algoritmasının Karmaşıklığı**

DFS'nin zaman karmaşıklığı, düğüm sayısı $(N)$ ve kenar sayısı $(E)$ ile doğru orantılıdır. Yani her düğümü ve her kenarı en az bir kez ziyaret eder. Bu nedenle zaman karmaşıklığı $O(N + E)$ olacaktır. Grafın kenar sayısı düğümlerden bağımsız olarak düşünüldüğünde, yalnızca düğüm sayısına göre $O(N)$ olarak da kabul edilebilir.

**Alan Karmaşıklığı**

 DFS, yığındaki eleman sayısına bağlı olarak bellek kullanır. En kötü durumda, graf tamamen tek bir yol şeklinde olduğunda, yığın en fazla N düğüm içerir. Bu yüzden alan karmaşıklığı $O(N)$'dir.

**Uygulama Alanları**

**Graflarda Yol Bulma:** DFS, özellikle belirli bir düğümden diğer düğüme gitmek için kullanılabilir. Eğer hedef düğüm bulunursa, arama işlemi sonlandırılır.

**Bağlantı Bileşenlerini Bulma:** DFS, graflarda bağlantı bileşenlerini bulmak için de kullanılabilir. Bir düğümden başlayarak, tüm bağlantılı düğümleri ziyaret eder.

**Ağaç Yapılarında Derinlemesine Arama:** DFS, özellikle ağaç yapılarında etkin bir şekilde kullanılır ve belirli bir yapıda en derine gitmeyi sağlar.

DFS, grafların keşfi ve analizi konusunda temel bir algoritma olarak öne çıkar. Bu algoritma, verilerin derinlemesine incelenmesi gereken durumlarda sıklıkla tercih edilir.

Şimdi algoritmamızı Romanya haritası üzerinde çalışalım.

```python
# DFS fonksiyonu başlangıç ve bitiş noktasıyla
def dfs(graph, start, goal, path=None, visited=None):
    if path is None:
        path = []
    if visited is None:
        visited = set()

    path.append(start)
    visited.add(start)

    # Hedefe ulaşıldı mı?
    if start == goal:
        return path

    # Komşuları gezme
    for neighbor in graph[start]:
        if neighbor not in visited:
            result = dfs(graph, neighbor, goal, path.copy(), visited)
            if result:  # Eğer hedef bulunduysa
                return result

    return None

# Romanya haritasını sözlük yapısına çevirelim

romania_map = {
    'Arad': ['Zerind', 'Timisoara', 'Sibiu'],
    'Zerind': ['Oradea', 'Arad'],
    'Oradea': ['Zerind', 'Sibiu'],
    'Timisoara': ['Arad', 'Lugoj'],
    'Lugoj': ['Timisoara', 'Mehadia'],
    'Mehadia': ['Lugoj', 'Drobeta'],
    'Drobeta': ['Mehadia', 'Craiova'],
    'Craiova': ['Drobeta', 'Rimnicu Vilcea', 'Pitesti'],
    'Rimnicu Vilcea': ['Sibiu', 'Craiova', 'Pitesti'],
    'Sibiu': ['Arad', 'Oradea', 'Fagaras', 'Rimnicu Vilcea'],
    'Fagaras': ['Sibiu', 'Bucharest'],
    'Pitesti': ['Rimnicu Vilcea', 'Craiova', 'Bucharest'],
    'Bucharest': ['Fagaras', 'Pitesti', 'Urziceni', 'Giurgiu'],
    'Giurgiu': ['Bucharest'],
    'Urziceni': ['Bucharest', 'Hirsova', 'Vaslui'],
    'Hirsova': ['Urziceni', 'Eforie'],
    'Eforie': ['Hirsova'],
    'Vaslui': ['Urziceni', 'Iasi'],
    'Iasi': ['Vaslui', 'Neamt'],
    'Neamt': ['Iasi']
}

# Başlangıç ve bitiş şehirleri
start_city = 'Arad'
goal_city = 'Bucharest'

# DFS algoritmasını çalıştırma
path = dfs(romania_map, start_city, goal_city)

if path:
    print(f"Seçilen {start_city} şehrinden varış noktası olan {goal_city} şehrine bulunan en kısa yol: {path}")
else:
    print(f"Seçilen şehirden {start_city} seçili şehire {goal_city} yol bulunmuyor.")
```

# **BFS**


BFS (Breadth-First Search), Türkçesiyle "Genişlik Öncelikli Arama", graflar üzerinde kullanılan en temel arama algoritmalarından biridir. BFS, grafın düğümlerini katman katman ziyaret eder, yani bir düğümden başlayıp o düğümün tüm komşularını ziyaret ettikten sonra komşuların komşularına geçer. Bu yaklaşım, derinliği önemsemeden her seviyeyi teker teker gezdiği için "genişlik öncelikli" olarak adlandırılır.

BFS, ağaçlarda, graflarda ve çeşitli veri yapılarında arama işlemi için sıklıkla tercih edilir. Özellikle iki düğüm arasındaki en kısa yolun bulunması gereken problemlerde çok etkilidir. Örneğin, bir şehirden başka bir şehre giden en kısa yolu bulmak ya da internet üzerinden iki kullanıcının birbirine kaç kişi uzaklıkta olduğunu hesaplamak gibi durumlarda kullanılabilir.

**BFS Nasıl Çalışır?**

BFS'nin çalışma mantığı, düğümleri sırayla ziyaret etmek için bir kuyruk (queue) veri yapısını kullanmasıdır. Bu yapı, "First In, First Out" (FIFO - İlk Giren İlk Çıkar) mantığına dayanır. BFS ile bir grafı taramak için şu adımlar izlenir:

1. **Başlangıç Düğümünü Seçme:**
Aramaya başlanacak bir başlangıç düğümü (node) seçilir.

2. **Kuyruğa Ekleme:**
Bu düğüm kuyruğa eklenir ve ziyaret edildiği işaretlenir.

3. **Komşuları Ziyaret Etme:**
Kuyruğun başındaki düğüm çıkarılır ve o düğümün komşuları ziyaret edilir. Ziyaret edilen düğümler kuyruğa eklenir.

4. **Tekrar Kontrol:**
Zaten ziyaret edilmiş düğümler tekrar kuyruğa eklenmez, böylece her düğüm yalnızca bir kez ziyaret edilir.

5. **Tüm Düğümler Ziyaret Edilene Kadar Devam Et:**
Kuyruk boşalana kadar bu işlem devam eder. Kuyruk boşaldığında, tüm düğümler genişlik sırasına göre ziyaret edilmiş olur.

##BFS'te Kuyruk Kullanımının Önemi

BFS (Genişlik Öncelikli Arama) algoritmasında kuyruk veri yapısının kullanılması, algoritmanın temel çalışma prensibi ile doğrudan ilişkilidir. Kuyruk, First In First Out (FIFO) yani "ilk giren ilk çıkar" ilkesine göre çalışan bir veri yapısıdır. Bu özellik, BFS'in genişlik öncelikli arama yapmasını sağlar.

##Neden Kuyruk?

Katman Katman İlerleme: BFS, bir grafın düğümlerini katmanlar halinde ziyaret eder. Yani, başlangıç düğümünden başlayarak birinci katmandaki tüm komşularını ziyaret eder, sonra bu komşuların komşularına (ikinci katman) geçer ve böylece devam eder. Kuyruk, bu katmanlı ilerlemeyi sağlamak için ideal bir yapıdır.

###FIFO Prensibi

Kuyrukta en önce eklenen eleman en önce çıkarılır. BFS'te de, öncelikle başlangıç düğümü kuyruğa eklenir. Daha sonra, bu düğümün komşuları kuyruğa eklenir. Kuyrukta en başta olan düğüm her zaman en önce ziyaret edilecek düğümdür. Bu sayede, BFS, en yakın komşuları önce ziyaret ederek genişliğe doğru ilerler.

###Ziyaret Edilmemiş Düğümleri Takip

Kuyruk, henüz ziyaret edilmemiş düğümleri tutar. Bu sayede, bir düğümün komşuları ziyaret edilirken, daha önce ziyaret edilmiş düğümlere tekrar dönülmez.

###En Kısa Yol Bulma

Ağırlıksız graflarda, BFS en kısa yolu bulmak için kullanılır. Kuyruk, her bir seviyedeki düğümleri sırayla ziyaret ederek, hedefe en kısa yoldaki düğümlere önce ulaşmayı sağlar.

###Kuyruk Olmadan Ne Olur?

Eğer BFS'te kuyruk yerine yığın gibi bir veri yapısı kullanırsak, algoritma DFS (Derinlik Öncelikli Arama) gibi davranmaya başlar. Yani, bir dalın en derin noktasına kadar inerek arama yapar. Bu da, BFS'in temel amacı olan genişliğe doğru arama yapma özelliğini kaybetmesine neden olur.

Sonuç olarak, BFS algoritmasında kuyruk veri yapısının kullanılması, algoritmanın genişlik öncelikli arama yapmasını sağlar. Kuyruğun FIFO özelliği, BFS'in en yakın komşuları önce ziyaret ederek katman katman ilerlemesini garanti eder. Bu sayede, BFS, özellikle ağırlıksız graflarda en kısa yolu bulma gibi birçok problemde etkili bir şekilde kullanılır.


##**BFS ve DFS Farkı**

**BFS**, genişliği önceleyerek grafın seviyelerini sırayla ziyaret eder. Bir düğümün tüm komşuları ziyaret edilmeden bir sonraki derinlikteki düğümlere geçilmez.

**DFS** (Depth-First Search) ise derinliği önceliklendirir. Bir düğümün komşularına derinlemesine giderek, düğümün sonuna kadar ilerler ve ancak sonra bir üst seviyeye döner.

**Kullanım Alanları**

**En Kısa Yol Bulma:** BFS, ağırlıksız graflarda iki düğüm arasındaki en kısa yolu bulmak için ideal bir algoritmadır.

**Graf Bileşenlerini Bulma:** Ayrık graflarda tüm bileşenleri bulmak için kullanılabilir.

**Labirent Çözme:** BFS, bir labirentte başlangıç noktasından çıkış noktasına en kısa yolu bulabilir.


**BFS'nin Zaman Karmaşıklığı:**

BFS algoritmasının zaman karmaşıklığı $O(V + E)$ şeklindedir. Burada $V$, grafın düğüm (vertex) sayısını, $E$ ise kenar (edge) sayısını ifade eder. Her düğüm ve her kenar bir kez işlenir, bu yüzden toplamda zaman karmaşıklığı düğüm ve kenar sayısına bağlıdır.

Yukardaki görselden daha net bir şekilde anlıyoruz ki BFS katman katman komşuları gezmekte aradaki mesafeyi dikkate almamaktadır.

```python
from collections import deque

# Romanya şehirleri grafını oluşturma
romania_map = {
    'Arad': ['Zerind', 'Timisoara', 'Sibiu'],
    'Zerind': ['Oradea', 'Arad'],
    'Oradea': ['Zerind', 'Sibiu'],
    'Timisoara': ['Arad', 'Lugoj'],
    'Lugoj': ['Timisoara', 'Mehadia'],
    'Mehadia': ['Lugoj', 'Drobeta'],
    'Drobeta': ['Mehadia', 'Craiova'],
    'Craiova': ['Drobeta', 'Rimnicu Vilcea', 'Pitesti'],
    'Rimnicu Vilcea': ['Sibiu', 'Craiova', 'Pitesti'],
    'Sibiu': ['Arad', 'Oradea', 'Fagaras', 'Rimnicu Vilcea'],
    'Fagaras': ['Sibiu', 'Bucharest'],
    'Pitesti': ['Rimnicu Vilcea', 'Craiova', 'Bucharest'],
    'Bucharest': ['Fagaras', 'Pitesti', 'Urziceni', 'Giurgiu'],
    'Giurgiu': ['Bucharest'],
    'Urziceni': ['Bucharest', 'Hirsova', 'Vaslui'],
    'Hirsova': ['Urziceni', 'Eforie'],
    'Eforie': ['Hirsova'],
    'Vaslui': ['Urziceni', 'Iasi'],
    'Iasi': ['Vaslui', 'Neamt'],
    'Neamt': ['Iasi']
}


def bfs(graph, start, goal):
    # Ziyaret edilen şehirler seti
    visited = set()

    # Kuyruğu başlatma ve başlangıç şehrini ekleme
    queue = deque([[start]])

    # Eğer başlangıç ve hedef şehir aynıysa
    if start == goal:
        return [start]

    while queue:
        # Kuyruğun ilk yolunu al
        path = queue.popleft()
        # Yoldaki son şehri al
        city = path[-1]

        # Eğer bu şehir daha önce ziyaret edilmediyse
        if city not in visited:
            neighbours = graph[city]

            # Komşulara bak
            for neighbour in neighbours:
                new_path = list(path)
                new_path.append(neighbour)
                queue.append(new_path)

                # Eğer hedefe ulaştıysak
                if neighbour == goal:
                    return new_path

            # Şehri ziyaret edilmiş olarak işaretle
            visited.add(city)

    return None



# Başlangıç ve bitiş şehirleri
start_city = 'Arad'
goal_city = 'Bucharest'

# DFS algoritmasını çalıştırma
path = bfs(romania_map, start_city, goal_city)


if path:
    print(f"Seçilen {start_city} şehrinden varış noktası olan {goal_city} şehrine bulunan en kısa yol: {path}")
else:
    print(f"Seçilen {start_city} şehrinden seçtiğiniz {goal_city} şehrine yol bulunmuyor.")
```

Bu sonuç, BFS algoritmasının genişlik öncelikli arama ile katman katman şehirleri gezerek en kısa yolu bulduğunu gösterir.

**BFS'nin Artıları:**

BFS, bir graf üzerinde iki düğüm arasındaki en kısa yolu bulmada mükemmeldir (ağırlıksız graflarda).

Ziyaret edilmemiş düğümleri kuyruk yapısıyla saklar, böylece her seferinde tüm komşuları inceler.

**BFS'nin Eksileri:**

Derin graf yapılarında BFS, çok fazla bellek tüketebilir. Özellikle geniş graflarda, kuyrukta çok sayıda düğüm saklanması gerekebilir.

BFS (Breadth-First Search) ve DFS (Depth-First Search) algoritmalarının zaman karmaşıklıkları aynı olsa da (her iki algoritmanın da zaman karmaşıklığı  $O(V+E)$, yani düğüm ve kenar sayısı kadar), alan karmaşıklıkları farklılık gösterir. Bu fark, her iki algoritmanın düğümleri ziyaret etme biçiminden kaynaklanır.

BFS'nin Alan Karmaşıklığı
BFS, kuyruk (queue) veri yapısını kullanarak grafiğin katman katman gezilmesini sağlar. BFS, bir düğümden başlayıp önce en yakın komşuları, sonra onların komşularını keşfeder. Bu süreçte, keşfedilen düğümler bir kuyruk yapısında saklanır ve bu kuyruk her seferinde genişler.

**Alan Karmaşıklığı (Space Complexity):**

**BFS'nin alan karmaşıklığı**

$O(V)$, yani en kötü durumda tüm düğümleri hafızada saklamak zorunda kalabilir. Çünkü BFS sırasında tüm düğümler kuyrukta tutulur ve her bir düğüm, komşularının hepsi işlenene kadar sırada bekler. Bir grafiğin en geniş seviyesi çok fazla düğüm içeriyorsa, o seviyedeki düğümlerin tümü aynı anda kuyrukta yer alacaktır. Özellikle geniş (çok fazla düğümü olan) ancak derin olmayan grafiklerde BFS'nin alan karmaşıklığı büyük olabilir. Bu nedenle, en kötü durumda BFS'nin alan karmaşıklığı $O(V)$ olur.


DFS ise yığın (stack) kullanarak grafikte derinlemesine araştırma yapar. Yani önce bir düğümden başlar ve mümkün olduğunca derinlemesine iner. Derine inebileceği başka düğüm kalmadığında geri döner ve bir sonraki komşuyu aramaya başlar. DFS'nin bellekte saklaması gereken veri, gidilen yolun izlenmesinden sorumlu olan yığın (stack) veri yapısıdır. Bu yığın, sadece derinlemesine inilen düğümleri saklar.

**DFS'in Alan Karmaşıklığı (Space Complexity)**

DFS, en kötü durumda en derin seviyeye kadar inebilir. Bu durumda, en derindeki düğüme ulaşana kadar yığında $O(h)$, yani ağacın veya grafiğin derinliği kadar düğüm saklanır. Eğer grafikteki en uzun yol ℎ ise, DFS'nin alan karmaşıklığı en kötü durumda
$O(h)$ olacaktır. Bu da, grafiğin yapısına göre çok daha düşük hafıza kullanımı anlamına gelebilir. Ancak, DFS'nin alan karmaşıklığı en kötü durumda grafiğin derinliği kadar olur, bu da O(V) olabilir, ancak genellikle derinliği
ℎ < < 𝑉 olan grafiklerde DFS, BFS'ye göre daha az bellek kullanır.

**BFS vs DFS Alan Karmaşıklığı Karşılaştırması**

BFS, geniş graflarda her bir katmandaki düğümleri hafızada tutmak zorunda olduğundan, en kötü durumda
𝑂 (𝑉) alan kullanır.
DFS, genellikle grafiğin derinliği kadar düğümü hafızada tutmak zorundadır, bu da derinliği ℎ olan bir grafikte 𝑂 (ℎ) alan anlamına gelir.


BFS'nin alan karmaşıklığı özellikle geniş ve yüzeysel grafikleri işlerken daha maliyetli olabilirken, DFS daha derin ancak dar grafiklerde avantajlı olabilir. Bu yüzden, grafiğin yapısına ve probleme bağlı olarak hangi algoritmanın kullanılacağı kararlaştırılmalıdır.

Bu farkları bilmek, BFS ve DFS'nin avantajları ve dezavantajları konusunda daha kapsamlı bir perspektif sunacaktır. Alan karmaşıklığı, özellikle bellek kısıtlamalarının olduğu durumlarda oldukça önemli hale gelir.

# **DIJKSTRA ALGORİTMASI**

Yukarıda gördüğümüz iki algoritma da ağırlıksız yol hesaplamaktaydı peki gerçek ağırlıklarla ve gerçek uzaklıklarla çalışacak olsaydık nasıl olurdu? Bu durumda da Dijkstra algoritması imdadımıza yetişiyor. Peki nedir Dijkstra algoritması?




Dijkstra Algoritması, bir kaynaktan (başlangıç düğümünden) diğer tüm düğümlere olan en kısa yolları bulmayı amaçlayan bir ağ algoritmasıdır. Bu algoritma, özellikle ağırlıklı graf yapılarında, negatif ağırlık içermeyen kenarlarda (yani mesafelerde) kullanılır. Ağırlıklar, genellikle mesafe, zaman veya maliyet gibi ölçümleri ifade eder.

**Dijkstra Algoritmasının Çalışma Mantığı**

Dijkstra algoritmasının temel amacı, bir grafın bir düğümünden (kaynak düğüm) diğer tüm düğümlere olan en kısa yolu bulmaktır. Algoritma, ağırlıkları olan kenarların bulunduğu bir graf üzerinde çalışır. Aşağıdaki temel adımlarla çalışır:

1. Başlangıç düğümünü belirle: Kaynak düğümünden başlar ve başlangıç düğümünün mesafesini 0 olarak ayarlar.

2. Diğer tüm düğümlerin mesafesini sonsuz yap: Henüz keşfedilmemiş düğümlere sonsuz mesafe atanır.

3. Komşu düğümleri güncelle: Her adımda, mevcut düğümden en yakın komşuya olan mesafeler hesaplanır. Eğer bu mesafe, daha önce o düğüme atanmış olan mesafeden daha kısa ise, bu düğümün mesafesi güncellenir.

4. Keşfedilen düğümleri işaretle: Bir düğüm işlenip komşularına olan mesafeleri güncellendikten sonra, bu düğüm işlenmiş (keşfedilmiş) olarak işaretlenir ve artık geri dönülmez.

5. Daha kısa bir yol bulunana kadar bu işlemi tekrarla: Tüm düğümler keşfedilene kadar işlem devam eder.

**Dijkstra Algoritmasının Karmaşıklığı**

**Zaman Karmaşıklığı**

Dijkstra algoritmasının zaman karmaşıklığı grafın veri yapısına ve kullanılan algoritma implementasyonuna göre değişir.
Ancak en iyi bilinen haliyle zaman karmaşıklığı, $O((V + E) log V)$

— burada V düğüm sayısını, E kenar sayısını temsil eder.

**Alan karmaşıklığı**

$O(V)$, çünkü her düğüm için en kısa yolu bulmak için ayrı bir mesafe saklanır.

Dijkstra, BFS'den farklı olarak ağırlıklı graf yapılarında çalışır. Ağırlıksız grafiklerde BFS en kısa yolu bulurken, Dijkstra ağırlıkları hesaba katar.

##PYTHON ile DIJKSTRA ALGORİTMASI UYGULAMASI

```python
import heapq


# Romanya haritası (Bu sefer uzaklıkları da hesaba katıyoruz)
romania_map = {
    'Oradea': {'Zerind': 71, 'Sibiu': 151},
    'Zerind': {'Oradea': 71, 'Arad': 75},
    'Arad': {'Zerind': 75, 'Sibiu': 140, 'Timisoara': 118},
    'Timisoara': {'Arad': 118, 'Lugoj': 111},
    'Lugoj': {'Timisoara': 111, 'Mehadia': 70},
    'Mehadia': {'Lugoj': 70, 'Drobeta': 75},
    'Drobeta': {'Mehadia': 75, 'Craiova': 120},
    'Craiova': {'Drobeta': 120, 'Rimnicu Vilcea': 146, 'Pitesti': 138},
    'Sibiu': {'Oradea': 151, 'Arad': 140, 'Fagaras': 99, 'Rimnicu Vilcea': 80},
    'Rimnicu Vilcea': {'Sibiu': 80, 'Pitesti': 97, 'Craiova': 146},
    'Pitesti': {'Rimnicu Vilcea': 97, 'Craiova': 138, 'Bucharest': 101},
    'Fagaras': {'Sibiu': 99, 'Bucharest': 211},
    'Bucharest': {'Fagaras': 211, 'Pitesti': 101, 'Urziceni': 85, 'Giurgiu': 90},
    'Giurgiu': {'Bucharest': 90},
    'Urziceni': {'Bucharest': 85, 'Hirsova': 98, 'Vaslui': 142},
    'Hirsova': {'Urziceni': 98, 'Eforie': 86},
    'Eforie': {'Hirsova': 86},
    'Vaslui': {'Urziceni': 142, 'Iasi': 92},
    'Iasi': {'Vaslui': 92, 'Neamt': 87},
    'Neamt': {'Iasi': 87}
}

def dijkstra(graph, start, goal):


    # Tüm düğümlerin başlangıç uzaklığını sonsuz olarak ayarlar
    distances = {node: float('inf') for node in graph}
    distances[start] = 0  # Başlangıç düğümünün uzaklığı 0

    # Her düğümün önceki düğümünü tutar (yolun geri izlenmesi için)
    previous_nodes = {node: None for node in graph}

    # Öncelik kuyruğu (mesafe, düğüm) şeklinde
    priority_queue = [(0, start)]

    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)

        # Eğer daha önce işlenmiş bir düğüm daha küçük mesafeye sahipse devam et
        if current_distance > distances[current_node]:
            continue

        # Hedef düğüme ulaşıldıysa yolu geri izle
        if current_node == goal:
            path = []
            while current_node is not None:
                path.insert(0, current_node)  # Yolu doğru sırada ekle
                current_node = previous_nodes[current_node]
            return distances[goal], path

        # Komşuları ziyaret et
        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                previous_nodes[neighbor] = current_node
                heapq.heappush(priority_queue, (distance, neighbor))

    return float('inf'), []  # Hedefe ulaşılamazsa sonsuz mesafe ve boş yol döndürülür

# Örnek kullanım
start_city = 'Arad'
goal_city = 'Bucharest'
distance, path = dijkstra(romania_map, start_city, goal_city)

print(f"{start_city} şehrinden {goal_city} şehrine olan en kısa mesafe: {distance} km")
print(f"İzlenen yol: {' -> '.join(path)}")
```

# **A* ( A-Star ) ALGORİTMASI**

A* algoritması (A-Star), graf veya yol bulma problemlerinde kullanılan etkili bir algoritmadır. A* hem uniform-cost search (en kısa yolu bulmaya odaklanan arama algoritması) hem de greedy search (hedefe en kısa sürede ulaşmayı amaçlayan arama) özelliklerini birleştirir. Bu sayede, yalnızca en kısa yolu bulmakla kalmaz, aynı zamanda hedefe ulaşma hızını da artırır. A* algoritması, maliyeti minimize etmek için $g(n) + h(n)$ şeklinde bir formül kullanır.

Burada:

$g(n)$: Başlangıç noktasından mevcut düğüme $(n)$ kadar olan gerçek maliyeti temsil eder.

$h(n)$: Heuristik fonksiyon, mevcut düğümden hedefe olan tahmini maliyeti belirtir.

Heuristik (Sezgisel) mesafe hesaplanmasında sorunun doğasına göre genellikle Öklidyen ya da Manhattan mesafe hesaplaması kullanılır.

##**A* Algoritmasının Çalışma Şekli**

1. **Başlangıç düğümünü (start node) açığa çıkarma**

Algoritmaya bir başlangıç düğümü verilir ve bu düğüm açığa alınır.

2. **Hedef düğümünü (goal node) bulma amacıyla ilerleme**

Algoritma, komşu düğümleri açığa çıkarırken her düğümün toplam maliyetini $(g + h)$ hesaplar.

3. **Açık listede en düşük maliyete sahip düğümü seçme**

Açık listede her düğümün maliyeti $(f(n) = g(n) + h(n))$ hesaplanır. En düşük maliyete sahip düğüm işleme alınır.


4. **Hedefe ulaşana kadar komşuları keşfetme**

Komşu düğümler keşfedilir, maliyetler hesaplanır ve düğümler sıralanır. Eğer düğüm hedefe ulaşıyorsa algoritma sonlandırılır.


5. **Yolu geriye doğru izleme**

Hedef düğüme ulaşıldığında, her düğümün "ebeveyn" düğümü üzerinden geri dönülerek en kısa yol elde edilir.


A* algoritması doğru seçilmiş bir heuristik fonksiyonla hedefe en kısa yolu bulabilir. Ancak, heuristik fonksiyon kötü seçilirse, algoritma performans kaybedebilir.

##**Zaman Karmaşıklığı**


**En kötü durumda:** $O(b^d)$, burada b dallanma faktörü, d ise hedefe olan mesafedir.

##**A* Algoritmasının Özellikleri**


**Tamlık:** Eğer hedef varsa, A* mutlaka bir çözüm bulur.


**Optimality:** Heuristik fonksiyon hedefe ulaşan yolun tahmininde doğruysa, A* en kısa yolu bulur.

##Heuristik Fonksiyonun Önemi

Heuristik fonksiyon, A* algoritmasının başarısı için kritik öneme sahiptir. İyi seçilmiş bir heuristik fonksiyon, algoritmanın daha hızlı çalışmasını sağlar. Ancak, heuristik fonksiyonun tahmini maliyetler her zaman gerçek maliyetten küçük olmalıdır (admissible). Aksi takdirde, algoritma optimal çözümü garanti edemez.

```python
import heapq

# Romanya şehirleri ve mesafeleri (heuristic fonksiyonu için kullanılan gerçek mesafeler)
romania_map = {
    'Oradea': {'Zerind': 71, 'Sibiu': 151},
    'Zerind': {'Oradea': 71, 'Arad': 75},
    'Arad': {'Zerind': 75, 'Sibiu': 140, 'Timisoara': 118},
    'Timisoara': {'Arad': 118, 'Lugoj': 111},
    'Lugoj': {'Timisoara': 111, 'Mehadia': 70},
    'Mehadia': {'Lugoj': 70, 'Drobeta': 75},
    'Drobeta': {'Mehadia': 75, 'Craiova': 120},
    'Craiova': {'Drobeta': 120, 'Rimnicu Vilcea': 146, 'Pitesti': 138},
    'Sibiu': {'Oradea': 151, 'Arad': 140, 'Fagaras': 99, 'Rimnicu Vilcea': 80},
    'Rimnicu Vilcea': {'Sibiu': 80, 'Pitesti': 97, 'Craiova': 146},
    'Pitesti': {'Rimnicu Vilcea': 97, 'Craiova': 138, 'Bucharest': 101},
    'Fagaras': {'Sibiu': 99, 'Bucharest': 211},
    'Bucharest': {'Fagaras': 211, 'Pitesti': 101, 'Urziceni': 85, 'Giurgiu': 90},
    'Giurgiu': {'Bucharest': 90},
    'Urziceni': {'Bucharest': 85, 'Hirsova': 98, 'Vaslui': 142},
    'Hirsova': {'Urziceni': 98, 'Eforie': 86},
    'Eforie': {'Hirsova': 86},
    'Vaslui': {'Urziceni': 142, 'Iasi': 92},
    'Iasi': {'Vaslui': 92, 'Neamt': 87},
    'Neamt': {'Iasi': 87}
}

# Heuristik fonksiyon (şehirlerin Bucharest'e olan tahmini mesafeleri)

heuristic = {
    'Arad': 366, 'Bucharest': 0, 'Craiova': 160, 'Drobeta': 242, 'Eforie': 161, 'Fagaras': 176,
    'Giurgiu': 77, 'Hirsova': 151, 'Iasi': 226, 'Lugoj': 244, 'Mehadia': 241, 'Neamt': 234,
    'Oradea': 380, 'Pitesti': 100, 'Rimnicu Vilcea': 193, 'Sibiu': 253, 'Timisoara': 329,
    'Urziceni': 80, 'Vaslui': 199, 'Zerind': 374
}

# A* algoritması
def a_star(graph, start, goal):
    open_list = []
    heapq.heappush(open_list, (0, start))
    came_from = {}
    g_score = {node: float('inf') for node in graph}
    g_score[start] = 0
    f_score = {node: float('inf') for node in graph}
    f_score[start] = heuristic[start]

    while open_list:
        current = heapq.heappop(open_list)[1]

        if current == goal:
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)
            path = path[::-1]
            total_cost = g_score[goal]
            print("Toplam maliyet (g(n)): ", total_cost)
            return path

        for neighbor, cost in graph[current].items():
            tentative_g_score = g_score[current] + cost
            if tentative_g_score < g_score[neighbor]:
                came_from[neighbor] = current
                g_score[neighbor] = tentative_g_score
                f_score[neighbor] = g_score[neighbor] + heuristic[neighbor]
                heapq.heappush(open_list, (f_score[neighbor], neighbor))

    return None

# Örnek: Arad'dan Bucharest'e en kısa yol ve maliyet hesaplaması
start_city = 'Arad'
goal_city = 'Bucharest'
path = a_star(romania_map, start_city, goal_city)
print("En kısa yol:", path)
```

Bir graph sorununda bir noktada başladığınız ya da geçtiğiniz noktaya yeniden dönüyorsanız orada döngü var demektir. Döngü tespiti, özellikle yönsüz graflarda önemlidir, çünkü döngüler bazen istenmeyen sonuçlara yol açabilir.

Döngü tespitinde iki yaygın yöntem vardır:

#**DFS (Depth-First Search - Derin Öncelikli Arama) Kullanarak Döngü Tespiti:**


##visited Kümesi ve Parent Dizisi Karşılaştırması

DFS (Derinlik Öncelikli Arama) algoritması, bir grafdaki tüm düğümleri ziyaret etmek için kullanılan güçlü bir yöntemdir. Aynı zamanda, graflarda döngü olup olmadığını tespit etmek için de kullanılabilir. Döngü tespiti için genellikle iki farklı yaklaşım kullanılır: visited kümesi ve parent dizisi.

###1. Visited Kümesi Yöntemi

Bir düğümü ziyaret ettiğimizde, "visited" kümesine ekleriz. Daha sonra, bir düğüme tekrar ulaştığımızda, bu düğümün zaten "visited" kümesinde olup olmadığını kontrol ederiz. Eğer kümede varsa, bu bir döngü olduğunu gösterir.

Avantajları: Basit ve anlaşılır bir yöntemdir.

Dezavantajları: Sadece döngü olup olmadığını söyler, döngünün nerede başladığını ve bittiğini belirtmez.

###2. Parent Dizisi Yöntemi

Her düğüm için bir "parent" (ebeveyn) düğüm tutarız. DFS sırasında bir düğüme ulaştığımızda, bu düğümün parent'ı ile kendimiz arasında bir kenar olup olmadığını kontrol ederiz. Eğer böyle bir kenar varsa, bu bir döngü olduğunu gösterir.

Avantajları: Döngünün nerede başladığını ve bittiğini belirlemek için kullanılabilir.

Dezavantajları: Ek bir dizi tutmak gerektiği için biraz daha fazla bellek kullanımı olabilir.

###Hangi Yöntemi Kullanmalıyız?

Sadece döngü olup olmadığını bilmek istiyorsak: visited kümesi yeterli olacaktır.
Döngünün nerede başladığını ve bittiğini bilmek istiyorsak: parent dizisi daha uygun olacaktır.


```python
def dfs_with_parent(graph, start, visited, parent):
    visited[start] = True
    for neighbor in graph[start]:
        if neighbor not in visited:
            parent[neighbor] = start
            if dfs_with_parent(graph, neighbor, visited, parent):
                return True  # Döngü bulundu
        elif neighbor != parent[start]:  # Kendini ebeveyni olarak işaretleme durumu hariç
            return True  # Döngü bulundu
    return False

# Kullanım örneği:
graph = {'A': ['B', 'C'], 'B': ['D', 'E'], 'C': ['B', 'F'], 'D': [], 'E': ['F'], 'F': []}
visited = {}
parent = {}
if dfs_with_parent(graph, 'A', visited, parent):
    print("Döngü var")
else:
    print("Döngü yok")
```

```python
def dfs_with_visited(graph, start, visited):
    visited.add(start)
    for neighbor in graph[start]:
        if neighbor not in visited:
            if dfs_with_visited(graph, neighbor, visited):
                return True  # Döngü bulundu
        else:
            return True  # Döngü bulundu
    return False

# Kullanım örneği:
graph = {'A': ['B', 'C'], 'B': ['D', 'E'], 'C': ['B', 'F'], 'D': [], 'E': ['F'], 'F': []}
visited = set()
if dfs_with_visited(graph, 'A', visited):
    print("Döngü var")
else:
    print("Döngü yok")
```

# **Union-Find Algoritması**

Union-Find algoritması, özellikle ayrık küme veri yapılarında ve döngü tespitinde oldukça kullanışlı bir araçtır. Şimdi bu algoritmayı daha detaylı inceleyelim.

##Union-Find Nedir?

Union-Find algoritması, bir nesne kümesini temsil etmek için kullanılan bir veri yapısıdır. Bu kümedeki nesneler başlangıçta birbirinden bağımsızdır. Algoritma, iki nesneyi aynı kümeye birleştirme (union) veya bir nesnenin hangi kümeye ait olduğunu bulma (find) işlemlerini gerçekleştirir.

##Çalışma Mantığı

Her bir nesne için bir kök düğüm: Başlangıçta her nesne kendi kök düğümüdür.

Ağaç yapısı: Union işlemleri sırasında, bir nesnenin kök düğümü başka bir nesnenin alt düğümü haline gelir ve böylece bir ağaç yapısı oluşur.

Path compression: Find işlemi sırasında, bir düğümün kök düğümüne ulaşmak için kullanılan yol üzerindeki tüm düğümlerin kök düğümünü doğrudan kök düğümüne işaret etmesi sağlanır. Bu sayede gelecekteki find işlemleri daha hızlı olur.

Rank (derinlik): Her düğüme bir rank (derinlik) değeri atanır. Union işlemi sırasında, rank'i daha küçük olan ağacın kökü, rank'i daha büyük olan ağacın köküne bağlanır. Bu sayede ağaçların dengeli kalması sağlanır ve algoritmanın performansı artırılır.

# **MINIMUM SPANNING TREE (MST)**
Bir graf üzerinde MST, tüm düğümleri bağlayan ve toplam kenar ağırlığı minimum olan bir alt ağaçtır. MST'yi bulmak için yaygın olarak kullanılan iki algoritma vardır: Kruskal ve Prim algoritmaları.

**MST'nin kullanıldığı bazı önemli alanlar**

# **1. Kruskal Algoritması**

Kruskal algoritması, "greedy" (açgözlü) bir yaklaşımla çalışır ve kenarları ağırlıklarına göre küçükten büyüğe sıralayarak minimum ağırlıklı bir ağaç oluşturur. Bu algoritma, Union-Find veri yapısını kullanarak döngüleri tespit eder.

##Algoritmanın Temel Mantığı

Kruskal algoritması, ağırlıklı bir grafi alır ve bu grafin tüm düğümlerini bağlayan en küçük toplam ağırlıklı ağacı (MST) bulur. Algoritmanın temel mantığı şu şekildedir:

1. Kenarları Sıralama: Tüm kenarlar ağırlıklarına göre artan sırada sıralanır.

2. Boş Bir Ağaç Oluşturma: Başlangıçta boş bir ağaç oluşturulur.

3. Kenarları Ekleme: Sıralanmış kenarlar tek tek incelenir. Eğer bir kenar, ağacı döngüye sokmadan birleştirirse, ağaca eklenir. Aksi takdirde atılır.

4. Döngü Kontrolü: Döngü kontrolü için genellikle Union-Find veri yapısı kullanılır.

**Kruskal Algoritmasının Çalışma Şekli**

Tüm kenarları ağırlıklarına göre küçükten büyüğe sırala.

En küçük ağırlıklı kenardan başlayarak, döngü oluşturmayacak şekilde kenarları ağaçta birleştir.

Tüm düğümler birbirine bağlanana kadar adım 2'yi tekrarla.

##PYTHON ile KRUSKAL ALGORİTMASININ UYGULANIŞI

```python
class UnionFind:
    def __init__(self, n):
        self.parent = list(range(n))
        self.rank = [0] * n

    def find(self, p):
        if self.parent[p] != p:
            self.parent[p] = self.find(self.parent[p])
        return self.parent[p]

    def union(self, p, q):
        rootP = self.find(p)
        rootQ = self.find(q)
        if rootP != rootQ:
            if self.rank[rootP] > self.rank[rootQ]:
                self.parent[rootQ] = rootP
            elif self.rank[rootP] < self.rank[rootQ]:
                self.parent[rootP] = rootQ
            else:
                self.parent[rootQ] = rootP
                self.rank[rootP] += 1

def kruskal_mst(graph):
    mst = []
    edges = sorted(graph['edges'], key=lambda x: x[2])  # Kenarları ağırlıklarına göre sırala
    uf = UnionFind(graph['vertices'])  # Union-Find yapısı

    for u, v, weight in edges:
        if uf.find(u) != uf.find(v):  # Döngü oluşturmayan kenarları ekle
            uf.union(u, v)
            mst.append((u, v, weight))

    return mst

graph = {
    'vertices': 6,
    'edges': [
        (0, 1, 4), (0, 2, 4), (1, 2, 2), (1, 3, 5), (2, 3, 5),
        (2, 4, 4), (3, 4, 5), (3, 5, 6), (4, 5, 7)
    ]
}

mst = kruskal_mst(graph)
print("Kruskal MST:", mst)
```

# **2. Prim Algoritması**

Prim algoritması da "greedy" bir algoritmadır. Ancak Kruskal'dan farklı olarak, düğümlerden başlayarak en küçük ağırlıklı kenarları seçer ve ağaçtaki mevcut düğümleri genişletir.

##Minimum Heap:
Prim algoritmasında kullanılan minimum heap, her zaman en küçük ağırlıklı kenarı bize verir. Bu sayede, her adımda ağaca en küçük maliyetle yeni bir düğüm eklenir.

##Ziyaret Edilen Düğümler:

visited listesi, hangi düğümlerin ağaca eklendiğini takip etmek için kullanılır. Bu sayede, aynı düğümü tekrar ekleme hatası yapılmaz.

##Maliyet:

Her düğüm için minimum heap'e eklenen maliyet, o düğüme ulaşmak için gereken en düşük maliyettir.

##Başlangıç Düğümü:

Prim algoritması, herhangi bir düğümden başlayabilir. Farklı başlangıç düğümleri farklı minimum spanning tree'ler verebilir, ancak toplam ağırlık her zaman aynı olacaktır.

##**Prim Algoritmasının Çalışma Şekli**

Bir başlangıç düğümü seç ve ağaca ekle.

Ağaçtaki düğümlere bağlı kenarların en küçük ağırlıklısını seç ve bu kenarı ağaca ekle.

Tüm düğümler eklenene kadar adım 2'yi tekrarla.

##PYTHON ile PRIM ALGORİTMASININ UYGULANIŞI

```python
import heapq

def prim_mst(graph):
    start_vertex = 0
    mst = []
    visited = [False] * len(graph)
    min_heap = [(0, start_vertex)]  # Başlangıç noktası, maliyet 0

    while min_heap:
        cost, u = heapq.heappop(min_heap)

        if visited[u]:
            continue

        visited[u] = True
        mst.append((u, cost))

        for v, weight in graph[u]:
            if not visited[v]:
                heapq.heappush(min_heap, (weight, v))

    return mst

graph = {
    0: [(1, 4), (2, 4)],
    1: [(0, 4), (2, 2), (3, 5)],
    2: [(0, 4), (1, 2), (3, 5), (4, 4)],
    3: [(1, 5), (2, 5), (4, 5), (5, 6)],
    4: [(2, 4), (3, 5), (5, 7)],
    5: [(3, 6), (4, 7)]
}

mst = prim_mst(graph)
print("Prim MST:", mst)
```

# **BELLMAN-FORD ALGORİTMASI**


Bellman-Ford algoritması, bir graf içinde bir başlangıç düğümünden tüm diğer düğümlere olan en kısa yolları bulmak için kullanılır. Bu algoritma, negatif ağırlıklı kenarları da işleyebilir, bu yönüyle Dijkstra’dan farklıdır. Ayrıca, graf içinde negatif ağırlıklı bir döngü (cycle) olup olmadığını da tespit edebilir.

**Bellman-Ford Algoritmasının Teorik Açıklaması**

**Girdi:** Ağırlıklı, yönlü bir grafik
𝐺 = (𝑉,𝐸) burada 𝑉 düğümler kümesi, 𝐸 ise kenarlar kümesidir. Her kenar (𝑢,𝑣) ile tanımlanır ve bu kenarın bir ağırlığı 𝑤(𝑢,𝑣) vardır.


**Çıktı:** Seçilen bir başlangıç düğümünden (source) tüm diğer düğümlere olan en kısa yol mesafeleri.

**Adımlar**

Başlangıç düğümünden diğer tüm düğümlere olan mesafeleri başlangıçta ∞ (sonsuz) olarak ayarlayın, başlangıç düğümünün kendisine olan mesafesini 0 yapın.
Her bir kenarı (𝑢,𝑣) için, 𝑑[𝑣] = min(𝑑[𝑣],𝑑[𝑢]+𝑤(𝑢,𝑣)) olacak şekilde güncelleme yapın. Bu adımı grafikteki düğüm sayısından ∣𝑉∣−1 kez tekrarlayın (düğüm sayısının bir eksiği kadar iterasyon yapın). Eğer bu adım sonrası bir kenarı tekrar rahatlatabiliyorsanız (yani mesafe daha da azalabiliyorsa), bu, grafikte negatif ağırlıklı bir döngü olduğunu gösterir.

##Rahatlatma (Relaxation):

Algoritmadaki "rahatlatma" işlemi, bir düğüme ulaşmak için daha kısa bir yol bulunursa, o düğüme olan mesafeyi güncelleme işlemidir. Bu işlem, en kısa yolu bulmak için iteratif olarak yapılır.

##Neden V-1 İterasyon?

Bellman-Ford algoritması, en fazla |V|-1 iterasyonda en kısa yolu bulur. Eğer grafikte negatif ağırlıklı döngü yoksa, |V|-1 iterasyondan sonra tüm düğümlere en kısa yol bulunmuş olur. Daha fazla iterasyon yapmanın bir anlamı yoktur.

##Negatif Ağırlıklı Döngü Tespiti:

Algoritmanın son adımında, tüm kenarları tekrar tarayarak herhangi bir rahatlatma işlemi yapılıp yapılmadığı kontrol edilir. Eğer bir rahatlatma işlemi yapılabilmişse, bu, grafikte negatif ağırlıklı bir döngü olduğu anlamına gelir.

##Dijkstra Algoritması ile Karşılaştırma:

Bellman-Ford algoritması, negatif ağırlıklı kenarları işleyebilmesi bakımından Dijkstra algoritmasından farklıdır. Dijkstra algoritması negatif ağırlıklı kenarlarda doğru sonuç vermez.


##**Zaman Karmaşıklığı:**

 Bellman-Ford’un zaman karmaşıklığı 𝑂(𝑉⋅𝐸) olup, grafikteki düğüm sayısı 𝑉, kenar sayısı ise 𝐸'dir.


##**Negatif Ağırlıklı Döngü Tespiti:**

Eğer bir kenarın ağırlığı bir adım daha rahatlatılabiliyorsa, bu, grafikte negatif ağırlıklı bir döngü bulunduğu anlamına gelir. Bu döngüler sonsuz bir şekilde döner ve en kısa yolu belirlemek mümkün olmaz.

```python
# Bellman-Ford algoritması
def bellman_ford(graph, start, end):
    # Şehirleri ve başlangıç noktasını ayarla
    distance = {node: float('inf') for node in graph}
    distance[start] = 0
    predecessor = {node: None for node in graph}

    # Relaxation işlemi, her düğüm için kenar sayısı kadar tekrar
    for _ in range(len(graph) - 1):
        for node in graph:
            for neighbor, weight in graph[node].items():
                if distance[node] + weight < distance[neighbor]:
                    distance[neighbor] = distance[node] + weight
                    predecessor[neighbor] = node

    # Negatif döngü kontrolü
    for node in graph:
        for neighbor, weight in graph[node].items():
            if distance[node] + weight < distance[neighbor]:
                print("Graf negatif ağırlıklı bir döngü içeriyor.")
                return None, None

    # Hedef şehre olan en kısa mesafeyi ve yolu bulmak
    path = []
    current_node = end
    while current_node is not None:
        path.append(current_node)
        current_node = predecessor[current_node]

    # Eğer path sadece end şehrini içeriyorsa, başlangıçtan oraya yol yoktur
    if len(path) == 1 and path[0] != start:
        return f"{start} şehrinden {end} şehrine ulaşılacak bir yol yok.", None

    path.reverse()  # Başlangıçtan bitişe doğru sıralamak için

    return distance[end], path

# Örnek kullanım (Arad'dan Bucharest'e olan en kısa yolu bulalım)
start_city = 'Arad'
end_city = 'Bucharest'
shortest_distance, path = bellman_ford(romania_map, start_city, end_city)

# Sonuçları yazdır
if path:
    print(f"{start_city} şehrinden {end_city} şehrine en kısa mesafe: {shortest_distance}")
    print(f"Geçilen şehirler: {path}")
else:
    print(shortest_distance)  # Ulaşılacak bir yol yoksa mesaj yazdır
```

# **FLOYD-WARSHALL ALGORİTMASI**

Floyd-Warshall algoritması, her düğümden her düğüme olan en kısa yolları bulan bir algoritmadır. Bellman-Ford gibi negatif ağırlıklı kenarlarla çalışabilir ve negatif ağırlıklı döngüler içeren grafikleri tespit eder. Floyd-Warshall, tüm düğümler arasındaki en kısa yolları bulduğu için farklı bir kullanım senaryosuna sahiptir.

Floyd-Warshall Algoritması Teorik Bilgi
Özellikler:

**Zaman Karmaşıklığı**

$𝑂(𝑉 ^ 3)$, burada 𝑉 düğüm sayısıdır.


**Alan Karmaşıklığı**

$𝑂(𝑉^2)$'dir, çünkü her düğüm çifti için mesafeleri saklayan bir matrise ihtiyaç duyar.

**Negatif ağırlıklar:**

Floyd-Warshall negatif ağırlıklı kenarlarla çalışabilir ancak negatif döngülerin varlığını tespit edebilir.

**Adım adım çalışma mantığı**

**1. Başlangıç Matrisinin Oluşturulması**

Grafikteki her düğüm çifti arasındaki mesafeyi bir matris halinde tutarız.
Başlangıçta, doğrudan bağlantı olan düğümler için kenar ağırlıkları, bağlantı olmayanlar için sonsuz (∞) değerler atanır. Her düğümden kendisine olan mesafe sıfırdır.

**2. Dinamik Güncelleme**

Üçlü bir döngü kullanarak her düğüm çifti arasında geçiş noktaları ekleyerek daha kısa yollar bulmaya çalışırız.
Düğüm 𝑘'yi bir ara geçiş düğümü olarak ekleyerek 𝑖 düğümünden 𝑗 düğümüne olan mesafeyi iyileştirip iyileştiremeyeceğimizi kontrol ederiz.

**3. Negatif Döngü Tespiti**

Eğer bir düğümün kendisine olan mesafesi negatif ise grafikte negatif döngü vardır.

```python
# Floyd-Warshall Algoritması
def floyd_warshall(graph):
    # Şehirleri ve başlangıç durumunu ayarla
    nodes = list(graph.keys())
    num_nodes = len(nodes)

    # Mesafe matrisi oluştur
    dist = {node: {node: float('inf') for node in nodes} for node in nodes}

    # Başlangıç mesafelerini yerleştir
    for node in graph:
        dist[node][node] = 0  # Kendi kendine olan mesafe 0'dır
        for neighbor, weight in graph[node].items():
            dist[node][neighbor] = weight

    # Floyd-Warshall algoritması
    for k in nodes:
        for i in nodes:
            for j in nodes:
                if dist[i][j] > dist[i][k] + dist[k][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]

    # Negatif döngü kontrolü
    for node in nodes:
        if dist[node][node] < 0:
            print(f"Negatif ağırlıklı bir döngü tespit edildi: {node}")
            return None

    return dist

# Mesafeleri daha okunabilir hale getiren fonksiyon
def print_distances(distances):
    for city1 in distances:
        for city2 in distances[city1]:
            if distances[city1][city2] == float('inf'):
                print(f"Mesafe {city1} ile {city2} arasında: Ulaşılamaz")
            else:
                print(f"Mesafe {city1} ile {city2} arasında: {distances[city1][city2]}")

# Romanya haritası
romania_map = {
    'Oradea': {'Zerind': 71, 'Sibiu': 151},
    'Zerind': {'Oradea': 71, 'Arad': 75},
    'Arad': {'Zerind': 75, 'Sibiu': 140, 'Timisoara': 118},
    'Timisoara': {'Arad': 118, 'Lugoj': 111},
    'Lugoj': {'Timisoara': 111, 'Mehadia': 70},
    'Mehadia': {'Lugoj': 70, 'Drobeta': 75},
    'Drobeta': {'Mehadia': 75, 'Craiova': 120},
    'Craiova': {'Drobeta': 120, 'Rimnicu Vilcea': 146, 'Pitesti': 138},
    'Sibiu': {'Oradea': 151, 'Arad': 140, 'Fagaras': 99, 'Rimnicu Vilcea': 80},
    'Rimnicu Vilcea': {'Sibiu': 80, 'Pitesti': 97, 'Craiova': 146},
    'Pitesti': {'Rimnicu Vilcea': 97, 'Craiova': 138, 'Bucharest': 101},
    'Fagaras': {'Sibiu': 99, 'Bucharest': 211},
    'Bucharest': {'Fagaras': 211, 'Pitesti': 101, 'Urziceni': 85, 'Giurgiu': 90},
    'Giurgiu': {'Bucharest': 90},
    'Urziceni': {'Bucharest': 85, 'Hirsova': 98, 'Vaslui': 142},
    'Hirsova': {'Urziceni': 98, 'Eforie': 86},
    'Eforie': {'Hirsova': 86},
    'Vaslui': {'Urziceni': 142, 'Iasi': 92},
    'Iasi': {'Vaslui': 92, 'Neamt': 87},
    'Neamt': {'Iasi': 87}
}

# Floyd-Warshall algoritmasını çalıştır
distances = floyd_warshall(romania_map)

# Sonuçları yazdır
if distances:
    print_distances(distances)
```

Yukarıda Floyd-Warshall algoritması ile bütün düğümlerin birbirine olan uzaklığının hesapladığı bir kod görmektesiniz şimdi bir de seçtiğimiz bir noktadan diğerine olan uzaklığın hesaplandığı bir kod yazalım.

```python
def floyd_warshall_path(graph):
    # Şehirleri ve başlangıç durumunu ayarla
    nodes = list(graph.keys())
    num_nodes = len(nodes)

    # Mesafe matrisi oluştur
    dist = {node: {node: float('inf') for node in nodes} for node in nodes}
    next_node = {node: {node: None for node in nodes} for node in nodes}

    # Başlangıç mesafelerini yerleştir
    for node in graph:
        dist[node][node] = 0  # Kendi kendine olan mesafe 0'dır
        for neighbor, weight in graph[node].items():
            dist[node][neighbor] = weight
            next_node[node][neighbor] = neighbor  # İlk adımda komşulara gideceğimiz yolu kaydediyoruz

    # Floyd-Warshall algoritması
    for k in nodes:
        for i in nodes:
            for j in nodes:
                if dist[i][j] > dist[i][k] + dist[k][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]
                    next_node[i][j] = next_node[i][k]

    return dist, next_node

# Seçilen iki şehir arasındaki en kısa yolu bulma
def get_path(next_node, start, end):
    if next_node[start][end] is None:
        return None
    path = [start]
    while start != end:
        start = next_node[start][end]
        path.append(start)
    return path

# Mesafeyi ve yolu yazdır
def print_shortest_path(dist, next_node, start, end):
    if dist[start][end] == float('inf'):
        print(f"{start} ve {end} arasında bir yol bulunmuyor.")
    else:
        path = get_path(next_node, start, end)
        print(f"En kısa mesafe {start} ile {end} arasında: {dist[start][end]} km")
        print(f"Yol: {' -> '.join(path)}")

# Romanya haritası
romania_map = {
    'Oradea': {'Zerind': 71, 'Sibiu': 151},
    'Zerind': {'Oradea': 71, 'Arad': 75},
    'Arad': {'Zerind': 75, 'Sibiu': 140, 'Timisoara': 118},
    'Timisoara': {'Arad': 118, 'Lugoj': 111},
    'Lugoj': {'Timisoara': 111, 'Mehadia': 70},
    'Mehadia': {'Lugoj': 70, 'Drobeta': 75},
    'Drobeta': {'Mehadia': 75, 'Craiova': 120},
    'Craiova': {'Drobeta': 120, 'Rimnicu Vilcea': 146, 'Pitesti': 138},
    'Sibiu': {'Oradea': 151, 'Arad': 140, 'Fagaras': 99, 'Rimnicu Vilcea': 80},
    'Rimnicu Vilcea': {'Sibiu': 80, 'Pitesti': 97, 'Craiova': 146},
    'Pitesti': {'Rimnicu Vilcea': 97, 'Craiova': 138, 'Bucharest': 101},
    'Fagaras': {'Sibiu': 99, 'Bucharest': 211},
    'Bucharest': {'Fagaras': 211, 'Pitesti': 101, 'Urziceni': 85, 'Giurgiu': 90},
    'Giurgiu': {'Bucharest': 90},
    'Urziceni': {'Bucharest': 85, 'Hirsova': 98, 'Vaslui': 142},
    'Hirsova': {'Urziceni': 98, 'Eforie': 86},
    'Eforie': {'Hirsova': 86},
    'Vaslui': {'Urziceni': 142, 'Iasi': 92},
    'Iasi': {'Vaslui': 92, 'Neamt': 87},
    'Neamt': {'Iasi': 87}
}

# Floyd-Warshall algoritmasını çalıştır
distances, next_nodes = floyd_warshall_path(romania_map)

# Örnek kullanım: Arad ile Bucharest arasındaki en kısa yol ve mesafe
start_city = 'Arad'
end_city = 'Bucharest'
print_shortest_path(distances, next_nodes, start_city, end_city)
```
