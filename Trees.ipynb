##**Binary Tree'nin KullanÄ±m AlanlarÄ±**

```python
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

# Binary Tree oluÅŸturma
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
```

Bu basit yapÄ±da bir Binary Tree oluÅŸturuyoruz:

KÃ¶k dÃ¼ÄŸÃ¼m 1 deÄŸerine sahip. KÃ¶kÃ¼n solunda 2 ve saÄŸÄ±nda 3 bulunuyor. Sol Ã§ocuk olan 2'nin solunda 4, saÄŸÄ±nda ise 5 var. Åimdi aÄŸaÃ§ yapÄ±sÄ±ndaki bu yapÄ±yÄ± gÃ¶rselleÅŸtirelim:

```python
#Bu fonksiyonlar farklÄ± sÄ±ralamalara gÃ¶re aÄŸaÃ§ Ã¼zerinde gezinti yapmanÄ± saÄŸlar.

# Pre-order traversal: KÃ¶k â†’ Sol â†’ SaÄŸ
def pre_order_traversal(root):
    if root:
        print(root.val, end=" ")
        pre_order_traversal(root.left)
        pre_order_traversal(root.right)

# In-order traversal: Sol â†’ KÃ¶k â†’ SaÄŸ
def in_order_traversal(root):
    if root:
        in_order_traversal(root.left)
        print(root.val, end=" ")
        in_order_traversal(root.right)

# Post-order traversal: Sol â†’ SaÄŸ â†’ KÃ¶k
def post_order_traversal(root):
    if root:
        post_order_traversal(root.left)
        post_order_traversal(root.right)
        print(root.val, end=" ")
```

##**1. Pre-order Traversal (Ã–nce KÃ¶k)**

Pre-order traversalâ€™da, dÃ¼ÄŸÃ¼mleri ÅŸu sÄ±raya gÃ¶re gezeriz:

KÃ¶k (root) dÃ¼ÄŸÃ¼mÃ¼nÃ¼ ziyaret et.

**AdÄ±mlar:**

Ã–nce kÃ¶k dÃ¼ÄŸÃ¼m (1) ziyaret edilir. Daha sonra sol Ã§ocuk (2) ziyaret edilir, ardÄ±ndan onun sol Ã§ocuÄŸu (4) ve saÄŸ Ã§ocuÄŸu (5). En son saÄŸ Ã§ocuk dÃ¼ÄŸÃ¼m (3) ziyaret edilir.

**Pre-order Traversal SÄ±rasÄ±: 1 â†’ 2 â†’ 4 â†’ 5 â†’ 3**

```python
def pre_order_traversal(root):
    if root:
        print(root.val, end=" ")
        pre_order_traversal(root.left)
        pre_order_traversal(root.right)

pre_order_traversal(root)
```

##**2. In-order Traversal (SÄ±ralÄ±)**

In-order traversalâ€™da, ÅŸu sÄ±raya gÃ¶re dÃ¼ÄŸÃ¼mleri ziyaret ederiz:

**In-order Traversal SÄ±rasÄ±: 4 â†’ 2 â†’ 5 â†’ 1 â†’ 3**

```python
def in_order_traversal(root):
    if root:
        in_order_traversal(root.left)
        print(root.val, end=" ")
        in_order_traversal(root.right)

in_order_traversal(root)
```

##**3. Post-order Traversal (Sonra KÃ¶k)**

Post-order traversalâ€™da ise ÅŸu sÄ±ra izlenir:

**Post-order Traversal SÄ±rasÄ±: 4 â†’ 5 â†’ 2 â†’ 3 â†’ 1**

```python
def post_order_traversal(root):
    if root:
        post_order_traversal(root.left)
        post_order_traversal(root.right)
        print(root.val, end=" ")

post_order_traversal(root)
```

##**4. Level-order Traversal (Seviye Seviyesi Gezinme)**

Level-order traversal, dÃ¼ÄŸÃ¼mleri aÄŸacÄ±n her seviyesini soldan saÄŸa doÄŸru ziyaret eder.

**AdÄ±mlar:**

Ä°lk olarak kÃ¶k dÃ¼ÄŸÃ¼m (1), ardÄ±ndan onun altÄ±ndaki sol (2) ve saÄŸ (3) dÃ¼ÄŸÃ¼mler ziyaret edilir. Sonra en alt seviyedeki dÃ¼ÄŸÃ¼mler, yani 4 ve 5, sÄ±rasÄ±yla ziyaret edilir.

**Level-order Traversal SÄ±rasÄ±: 1 â†’ 2 â†’ 3 â†’ 4 â†’ 5**

```python
from collections import deque

def level_order_traversal(root):
    if not root:
        return
    queue = deque([root])
    while queue:
        node = queue.popleft()
        print(node.val, end=" ")
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)

level_order_traversal(root)
```

AVL aÄŸacÄ±, her dÃ¼ÄŸÃ¼mde sol ve saÄŸ alt aÄŸaÃ§larÄ±n yÃ¼kseklik farkÄ±nÄ±n en fazla 1 olmasÄ±nÄ± garanti eden bir dengeli ikili arama aÄŸacÄ± tÃ¼rÃ¼dÃ¼r. Her dÃ¼ÄŸÃ¼mdeki bu fark, dÃ¼ÄŸÃ¼mÃ¼n denge faktÃ¶rÃ¼ (balance factor) olarak adlandÄ±rÄ±lÄ±r ve ÅŸu ÅŸekilde hesaplanÄ±r:

##**AVL AÄŸacÄ±nda Rotasyonlar**

AVL aÄŸacÄ±nda dengesizlik oluÅŸtuÄŸunda, dÃ¶rt farklÄ± durum iÃ§in rotasyon uygulanÄ±r. Bu rotasyonlar ÅŸu ÅŸekildedir;

###**1. LL (Left-Left) Rotasyonu:**
EÄŸer bir dÃ¼ÄŸÃ¼mÃ¼n sol alt aÄŸacÄ± aÄŸÄ±rsa (yani, denge faktÃ¶rÃ¼ 2) ve sol Ã§ocuÄŸun sol alt aÄŸacÄ± da aÄŸÄ±rsa, bu durumda LL rotasyonu yapÄ±lÄ±r.

###**2. RR (Right-Right) Rotasyonu:**
EÄŸer bir dÃ¼ÄŸÃ¼mÃ¼n saÄŸ alt aÄŸacÄ± aÄŸÄ±rsa (denge faktÃ¶rÃ¼ -2) ve saÄŸ Ã§ocuÄŸun saÄŸ alt aÄŸacÄ± da aÄŸÄ±rsa, bu durumda RR rotasyonu yapÄ±lÄ±r.

SaÄŸ alt aÄŸaÃ§ tek taraflÄ± bÃ¼yÃ¼dÃ¼ÄŸÃ¼nde, sola doÄŸru tek rotasyon yapÄ±lÄ±r.

###**3. LR (Left-Right) Rotasyonu:**
EÄŸer sol alt aÄŸaÃ§ aÄŸÄ±rsa (denge faktÃ¶rÃ¼ 2) fakat sol Ã§ocuÄŸun saÄŸ alt aÄŸacÄ± aÄŸÄ±rsa, bu durumda LR rotasyonu yapÄ±lÄ±r.

Sol-alt-saÄŸ bÃ¼yÃ¼mesi olduÄŸunda, Ã¶nce sola kÃ¼Ã§Ã¼k bir rotasyon yapÄ±lÄ±r (LL), ardÄ±ndan saÄŸa bir rotasyon yapÄ±lÄ±r.

###**4. RL (Right-Left) Rotasyonu:**
EÄŸer saÄŸ alt aÄŸaÃ§ aÄŸÄ±rsa (denge faktÃ¶rÃ¼ -2) fakat saÄŸ Ã§ocuÄŸun sol alt aÄŸacÄ± aÄŸÄ±rsa, bu durumda RL rotasyonu yapÄ±lÄ±r.

SaÄŸ-alt-sol bÃ¼yÃ¼mesi olduÄŸunda, Ã¶nce saÄŸa kÃ¼Ã§Ã¼k bir rotasyon yapÄ±lÄ±r (RR), ardÄ±ndan sola bir rotasyon yapÄ±lÄ±r.

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None
        self.height = 1  # YÃ¼ksekliÄŸi takip etmek iÃ§in her dÃ¼ÄŸÃ¼mde 'height' tutulur

class AVLTree:
    def insert(self, root, key):
        # Normal BST ekleme iÅŸlemi
        if not root:
            return Node(key)
        elif key < root.key:
            root.left = self.insert(root.left, key)
        else:
            root.right = self.insert(root.right, key)

        # YÃ¼ksekliÄŸi gÃ¼ncelle
        root.height = 1 + max(self.get_height(root.left), self.get_height(root.right))

        # Denge faktÃ¶rÃ¼nÃ¼ hesapla
        balance = self.get_balance(root)

        # Dengesizlik durumlarÄ±
        # LL Durumu
        if balance > 1 and key < root.left.key:
            return self.right_rotate(root)

        # RR Durumu
        if balance < -1 and key > root.right.key:
            return self.left_rotate(root)

        # LR Durumu
        if balance > 1 and key > root.left.key:
            root.left = self.left_rotate(root.left)
            return self.right_rotate(root)

        # RL Durumu
        if balance < -1 and key < root.right.key:
            root.right = self.right_rotate(root.right)
            return self.left_rotate(root)

        return root

    # Sol rotasyon iÅŸlemi
    def left_rotate(self, z):
        y = z.right
        T2 = y.left

        # Rotasyon iÅŸlemi
        y.left = z
        z.right = T2

        # YÃ¼kseklikleri gÃ¼ncelle
        z.height = 1 + max(self.get_height(z.left), self.get_height(z.right))
        y.height = 1 + max(self.get_height(y.left), self.get_height(y.right))

        return y

    # SaÄŸ rotasyon iÅŸlemi
    def right_rotate(self, z):
        y = z.left
        T3 = y.right

        # Rotasyon iÅŸlemi
        y.right = z
        z.left = T3

        # YÃ¼kseklikleri gÃ¼ncelle
        z.height = 1 + max(self.get_height(z.left), self.get_height(z.right))
        y.height = 1 + max(self.get_height(y.left), self.get_height(y.right))

        return y

    # YÃ¼kseklik hesaplama
    def get_height(self, node):
        if not node:
            return 0
        return node.height

    # Denge faktÃ¶rÃ¼nÃ¼ hesaplama
    def get_balance(self, node):
        if not node:
            return 0
        return self.get_height(node.left) - self.get_height(node.right)

    # Inorder traversal (aÄŸaÃ§ iÃ§eriÄŸini yazdÄ±rma)
    def inorder(self, root):
        if not root:
            return
        self.inorder(root.left)
        print(f"{root.key} ", end="")
        self.inorder(root.right)

# AVL aÄŸacÄ±nÄ± test etme
if __name__ == "__main__":
    tree = AVLTree()
    root = None

    # ElemanlarÄ± ekleme
    numbers = [10, 20, 30, 40, 50, 25]
    for number in numbers:
        root = tree.insert(root, number)

    # AÄŸaÃ§ iÃ§eriÄŸini yazdÄ±rma
    print("AVL Tree'nin inorder traversal yazÄ±lÄ±mÄ±: ", end="")
    tree.inorder(root)
    print()
```

##**3. Heap Sort (Heap SÄ±ralama)**

Heap kullanÄ±larak veri sÄ±ralamasÄ± yapÄ±labilir. Bu yÃ¶ntem Heap Sort olarak adlandÄ±rÄ±lÄ±r ve ÅŸu adÄ±mlarÄ± iÃ§erir:

KÃ¶k dÃ¼ÄŸÃ¼mÃ¼ (en bÃ¼yÃ¼k eleman) alÄ±nÄ±r ve dizinin sonuna konur.

Yeni kÃ¶k dÃ¼ÄŸÃ¼mÃ¼ olarak son eleman alÄ±nÄ±r ve yeniden heapify-down iÅŸlemi uygulanarak heap yeniden yapÄ±landÄ±rÄ±lÄ±r.

##**4. Heap'in Array ile Temsili**

Heap'ler, verimli bir ÅŸekilde array (dizi) ile temsil edilebilir. Bu temsilde:

KÃ¶k dÃ¼ÄŸÃ¼mÃ¼ dizi baÅŸÄ±nda bulunur: A[0]

Bir dÃ¼ÄŸÃ¼mÃ¼n sol Ã§ocuÄŸu $2 * i + 1$, saÄŸ Ã§ocuÄŸu ise $2 * i + 2$ indeksinde bulunur.

Bir dÃ¼ÄŸÃ¼mÃ¼n ebeveyni ise $(i - 1) / 2$ indeksindedir.







```python
class MaxHeap:
    def __init__(self):
        self.heap = []

    def parent(self, i):
        return (i - 1) // 2

    def leftChild(self, i):
        return 2 * i + 1

    def rightChild(self, i):
        return 2 * i + 2

    def insert(self, key):
        self.heap.append(key)
        self._heapify_up(len(self.heap) - 1)

    def delete(self):
        if len(self.heap) == 0:
            return None
        if len(self.heap) == 1:
            return self.heap.pop()

        root = self.heap[0]
        self.heap[0] = self.heap.pop()  # Son elemanÄ± kÃ¶ke taÅŸÄ±yoruz
        self._heapify_down(0)  # Heap'i aÅŸaÄŸÄ± doÄŸru yeniden dÃ¼zenliyoruz
        return root

    def _heapify_up(self, index):
        # YukarÄ± doÄŸru hareket ederek heap'in dengesini saÄŸlÄ±yoruz
        while index != 0 and self.heap[self.parent(index)] < self.heap[index]:
            self.heap[self.parent(index)], self.heap[index] = self.heap[index], self.heap[self.parent(index)]
            index = self.parent(index)

    def _heapify_down(self, index):
        # AÅŸaÄŸÄ± doÄŸru hareket ederek heap'in dengesini saÄŸlÄ±yoruz
        largest = index
        left = self.leftChild(index)
        right = self.rightChild(index)

        if left < len(self.heap) and self.heap[left] > self.heap[largest]:
            largest = left

        if right < len(self.heap) and self.heap[right] > self.heap[largest]:
            largest = right

        if largest != index:
            self.heap[index], self.heap[largest] = self.heap[largest], self.heap[index]
            self._heapify_down(largest)

    def get_max(self):
        if len(self.heap) == 0:
            return None
        return self.heap[0]

# KullanÄ±m:
heap = MaxHeap()
heap.insert(10)
heap.insert(20)
heap.insert(15)
heap.insert(30)
print("Max Eleman:", heap.get_max())  # Max Eleman: 30
print("KÃ¶k Eleman Silindi:", heap.delete())  # KÃ¶k Eleman Silindi: 30
print("Yeni Max Eleman:", heap.get_max())  # Yeni Max Eleman: 20
```

Bir kelimeyi Trie'de ararken, kelimenin her karakterini kÃ¶kten baÅŸlayarak ararÄ±z. EÄŸer karakterlerin tamamÄ±nÄ± bulursak ve kelimenin sonunu iÅŸaret eden bir dÃ¼ÄŸÃ¼m varsa, kelime Trie'de vardÄ±r.

**c. Silme (Delete)**

Bir kelimeyi silerken, Ã¶nce kelimenin Trie'de olup olmadÄ±ÄŸÄ±nÄ± kontrol ederiz. EÄŸer kelime Trie'de varsa, kelimenin dÃ¼ÄŸÃ¼mlerini silmeye baÅŸlarÄ±z. EÄŸer dÃ¼ÄŸÃ¼mÃ¼n baÅŸka bir kelimenin parÃ§asÄ± olmadÄ±ÄŸÄ± belirlenirse, bu dÃ¼ÄŸÃ¼m ve alt dÃ¼ÄŸÃ¼mleri silinir.

```python
class TrieNode:
    def __init__(self):
        self.children = {}  # Ã‡ocuk dÃ¼ÄŸÃ¼mleri depolamak iÃ§in bir sÃ¶zlÃ¼k kullanÄ±lÄ±r
        self.is_end_of_word = False  # Bu dÃ¼ÄŸÃ¼mÃ¼n bir kelimenin sonu olup olmadÄ±ÄŸÄ±nÄ± gÃ¶sterir

class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        """
        Trie'ye bir kelime ekler.
        """
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()  # EÄŸer karakter yoksa, yeni bir dÃ¼ÄŸÃ¼m oluÅŸtur
            node = node.children[char]
        node.is_end_of_word = True  # Kelimenin sonuna gelindiÄŸinde iÅŸaretlenir
        print(f"{word}")

    def search(self, word):
        """
        Trie'de bir kelime arar. Kelime varsa True, yoksa False dÃ¶ner.
        """
        node = self.root
        for char in word:
            if char not in node.children:
                return False  # EÄŸer karakter yoksa, kelime de yoktur
            node = node.children[char]
        return node.is_end_of_word  # Son dÃ¼ÄŸÃ¼mde kelimenin sonu olup olmadÄ±ÄŸÄ±na bakÄ±lÄ±r


    def starts_with(self, prefix):
        """
        Verilen Ã¶nekle baÅŸlayan kelimelerin olup olmadÄ±ÄŸÄ±nÄ± kontrol eder.
        """
        node = self.root
        for char in prefix:
            if char not in node.children:
                return False  # Ã–nek bulunamazsa False dÃ¶ner
            node = node.children[char]
        return True  # EÄŸer Ã¶nek bulunursa True dÃ¶ner

    def delete(self, word):
        """
        Trie'den bir kelimeyi siler.
        """
        def _delete(node, word, depth):
            # TÃ¼m kelime iÅŸlendiyse ve bu bir kelimenin sonuysa, iÅŸaretlemeyi kaldÄ±r
            if depth == len(word):
                if not node.is_end_of_word:
                    return False  # Kelime Trie'de bulunmuyorsa
                node.is_end_of_word = False
                return len(node.children) == 0  # EÄŸer yaprak dÃ¼ÄŸÃ¼mse silmek iÃ§in True dÃ¶ner

            # RekÃ¼rsif olarak dÃ¼ÄŸÃ¼mleri silmek iÃ§in Ã§aÄŸrÄ±
            char = word[depth]
            if char not in node.children:
                return False  # EÄŸer kelime Trie'de bulunmuyorsa

            should_delete_current_node = _delete(node.children[char], word, depth + 1)

            # Ã‡ocuk dÃ¼ÄŸÃ¼mÃ¼ sil
            if should_delete_current_node:
                del node.children[char]
                return len(node.children) == 0

            return False

        _delete(self.root, word, 0)

# KullanÄ±m:
trie = Trie()
trie.insert("apple")
print(trie.search("apple"))  # True
print(trie.search("app"))  # False
print(trie.starts_with("app"))  # True
trie.insert("app")
print(trie.search("app"))  # True
trie.delete("apple")
print(trie.search("apple"))  # False
```

##**Fenwick Tree'nin temel Ã¶zellikleri**

##**Fenwick Tree Teorisi**

Fenwick Tree, bir dizinin elemanlarÄ±na dayalÄ± bir yapÄ±dÄ±r ve dizinin her indeksine bir ikili sayÄ± gÃ¶zÃ¼yle bakÄ±lÄ±r. Dizideki her bir elemanÄ±n toplamÄ±, en saÄŸdaki 1'den sonraki tÃ¼m bitleri kapsayan alt aralÄ±ÄŸÄ±n toplamÄ±nÄ± temsil eder.

Her gÃ¼ncelleme iÅŸleminde, ilgili indeksin Ã¼st aralÄ±klarÄ±na eklemeler yapÄ±lÄ±r. Benzer ÅŸekilde, sorgulama yaparken daha geniÅŸ alt aralÄ±klarÄ±n toplamlara dahil edilmesi iÃ§in indeksler azaltÄ±larak ilerlenir.

##**Fenwick Tree'nin AÅŸamalarÄ±**

**GÃ¼ncelleme:** Dizideki bir deÄŸeri artÄ±rmak ya da azaltmak.

**Python Koduyla Fenwick Tree (BIT) UygulamasÄ±**

AÅŸaÄŸÄ±da Fenwick Tree'yi (Binary Indexed Tree) Python'da nasÄ±l uygulayabileceÄŸimizi adÄ±m adÄ±m aÃ§Ä±klayan bir kod Ã¶rneÄŸi bulunmaktadÄ±r:

```python
class FenwickTree:
    def __init__(self, n):
        """Fenwick Tree oluÅŸturma. Boyut n olan bir diziyle Ã§alÄ±ÅŸacaÄŸÄ±z."""
        self.n = n
        self.tree = [0] * (n + 1)  # 1'den indekslenen dizimiz

    def update(self, index, value):
        """index numaralÄ± elemanÄ± value kadar gÃ¼ncelle."""
        while index <= self.n:
            self.tree[index] += value
            index += index & -index  # SaÄŸdaki 1'i bul ve ekle

    def query(self, index):
        """1'den index'e kadar olan toplamÄ± dÃ¶ndÃ¼r."""
        result = 0
        while index > 0:
            result += self.tree[index]
            index -= index & -index  # SaÄŸdaki 1'i Ã§Ä±kar ve bir Ã¶nceki alt gruba geÃ§
        return result

    def range_query(self, left, right):
        """left ile right arasÄ±ndaki toplamÄ± dÃ¶ndÃ¼r."""
        return self.query(right) - self.query(left - 1)

# Ã–rnek kullanÄ±m
if __name__ == "__main__":
    # 8 elemanlÄ± bir dizimiz var.
    ft = FenwickTree(8)

    # Dizimize bazÄ± deÄŸerler ekleyelim.
    values = [0, 3, 2, -1, 6, 5, 4, -3, 3]  # 1'den indekslenmiÅŸ
    for i in range(1, len(values)):
        ft.update(i, values[i])

    # Sorgulamalar yapalÄ±m
    print("1'den 5'e kadar olan toplam:", ft.query(5))  # 1'den 5'e kadar olan toplam
    print("3 ile 7 arasÄ±ndaki toplam:", ft.range_query(3, 7))  # 3'ten 7'ye kadar olan toplam
```

**update metodu:** Bir indeksteki deÄŸeri gÃ¼ncellemek iÃ§in kullanÄ±lÄ±r. Ä°ndeksin ikili gÃ¶sterimindeki en saÄŸdaki 1 bitini kullanarak ilgili aralÄ±klarÄ± gÃ¼ncelleriz.

**query metodu:** 1'den belirtilen indeks numarasÄ±na kadar olan toplamÄ± dÃ¶ndÃ¼rÃ¼r. Ä°ndeks ikili gÃ¶steriminde saÄŸdaki 1 bitini kullanarak toplama iÅŸlemi yapar.
