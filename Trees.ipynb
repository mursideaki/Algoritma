# **BINARY TREE (İkili Ağaç)**

Bir Binary Tree, her düğümün (node) en fazla iki alt düğüme sahip olduğu bir veri yapısıdır. Bu alt düğümler "sol çocuk" (left child) ve "sağ çocuk" (right child) olarak adlandırılır. Bu yapı, verileri hiyerarşik bir düzende depolamayı sağlar.


##**Terminoloji**

**Kök (Root):** Ağacın en üst düğümüdür yani ağacın başlangıç noktasıdır.

**Düğüm (Node):** Veriyi saklayan her bir birimdir. Kendi sol ve/veya sağ çocukları olabilir.

**Yaprak (Leaf):** Hiç çocuk düğümü olmayan düğümlere yaprak denir. Ağacın en alt kısmında yer alırlar.

**Alt Ağaç (Subtree):** Bir düğüm ve o düğümün tüm alt düğümleri bir alt ağaç oluşturur.

**Yükseklik (Height):** Bir ağacın kökten en uzak yaprağa kadar olan en uzun yolu ifade eder.

**Derinlik (Depth):** Bir düğümün kökten olan uzaklığıdır.

###**İkili Ağaçların Temel Özellikleri**

Her düğüm maksimum iki çocuk düğüme sahiptir. Bir düğüm yeni çocuk düğümlerle dallara ayrılabilir, bu da veri hiyerarşisini sağlar. Çocuk düğümü olmayan düğümlere yaprak denir. Yapraklar ağaç yapılarının en alt düzeyindedirler. Binary treelerde döngü olamaz yani kökten yaprağa tek bir yol bulunur. Binary treelerde tek bir root bulunmak zorundadır.


###**Düzey ve Düğüm Sayısı**

İkili ağaçta kök düğümün düzeyini (level) 0 olarak kabul edersek, düzey
𝑑 için düğüm sayısı $2^𝑑$ olarak hesaplanır. Bu, her düzeyde mevcut düğümlerin iki katına çıktığını gösterir.

**Örnek:** Düzey 3’teki maksimum düğüm sayısı
$2^3 = 8$ olduğundan, düzey 3’te en fazla 8 düğüm olabilir.

###**Yükseklik ve Düğüm Sayısı**

Kök düğümün derinliğini (depth) 1 olarak kabul ettiğimizde, bir ikili ağacın yüksekliği ℎ için maksimum düğüm sayısı $2^ℎ-1$ olarak belirlenir. Bu formül, tam dolu bir ikili ağaç için geçerlidir.

Örnek: Yüksekliği 5 olan bir ikili ağaçta toplam düğüm sayısı;
$2^5−1=31$ olur.

###**Minimum Yükseklik**

𝑁 düğümü bulunan bir ikili ağacın minimum yüksekliği ve düzeyi
$log₂(N+1)$ olarak hesaplanır. Burada çıkan sonuç tam sayı olarak yukarı yuvarlanır.

**Örnek:**
Düğüm sayısı 6 olan bir ağacın minimum yüksekliği
$log₂(6+1)=3$ olur.

###**Yaprak Düğüm Sayısı**

Kök düğümün düzeyini 1 olarak kabul edersek, 𝐿 yaprak sayısına sahip bir ikili ağacın düzeyi en az $log₂L+1$ olur. Çıkan sayı yine tam sayı olarak yukarı yuvarlanır.

Örnek: Yaprak sayısı 6 olan bir ağacın minimum düzeyi
$log₂6+1=4$ olur.

###**Yaprak Düğüm İlişkisi**

Her düğümün ya 0 ya da 2 çocuğa sahip olduğu bir ikili ağaçta, yaprak düğüm sayısı her zaman iki çocuklu düğümlerin bir fazlasıdır. Bu durum, ağacın yapısının nasıl organize edildiğini gösterir.

# **İKİLİ AĞAÇ TÜRLERİ**

##**Full Binary Tree**

Her düğümün ya 0 ya da 2 çocuğa sahip olduğu ağaç türüdür. Yaprak düğümler dışındaki her düğümde iki çocuk bulunur. Bu durum, tam ikili ağaçların en belirgin özelliklerindendir.

Düğüm sayısı (n), yaprak sayısı (y) ve dahili düğüm sayısı (i) arasındaki ilişkiler şöyledir;

Yaprak sayısı $(y) = i+ 1$

Toplam düğüm sayısı $(n) = 2i + 1$


##**Complete Binary Tree**

Son düzeyi hariç diğer tüm düzeyleri tam dolu ve son düzeyi mümkün olduğu kadar doldurulmuş ağaçtır. Yaprak düğümleri sol taraftan doldurulur.
Tam ikili ağaçlardan farklı olarak, son düzeyde sağda boş alan kalabilir.


##**Perfect Binary Tree**

Her yaprak düğümün aynı seviyede olduğu ve son düzey hariç her düzeydeki düğümlerin iki çocuğa sahip olduğu ağaç türüdür.

Yüksekliği ℎ olan bir Perfect Binary Tree’nin düğüm sayısı $2ʰ⁺¹ - 1$ olarak hesaplanır.

##**Balanced Binary Tree**

Her bir düğüm için sol ve sağ alt ağaçların yüksekliği arasındaki farkın sıfır veya bir olduğu ağaç türüdür. Bu denge, ağacın işlemlerinin verimliliğini artırır.

Dengesiz bir ağaçta, bir düğümün sol ve sağ alt ağaçlarının yüksekliği arasında büyük bir fark olabilir, bu da bazı işlemlerin yavaşlamasına neden olabilir.

##**Dengeli ve Dengesiz Ağaçlar**

Bir dengesiz ağaçta, belirli bir düğümün alt ağaçlarının yüksekliği arasında büyük farklar olabilir. Bu durumda, örneğin bir düğümün sol alt ağacı yüksek iken sağ alt ağaç hiç var olmayabilir. Bu, ağacın dengesiz olduğunu gösterir.
Dengeleyici ağaçlarda, her düğümün sol ve sağ alt ağaçları arasında en fazla 1 seviyelik bir fark bulunur. Bu, ağacın daha verimli işlemesini sağlar.


##**Binary Tree Traversal (Ağaç Gezinme):**
Bir binary tree üzerinde gezinmek, veriyi farklı sıralarda işlemeyi sağlar.

#**Gezinme yöntemleri**

Pre-order Traversal (Önce Kök): Kök → Sol → Sağ

In-order Traversal (Sıralı): Sol → Kök → Sağ

Post-order Traversal (Sonra Kök): Sol → Sağ → Kök

Level-order Traversal (Seviye Seviyesine): Her seviyeyi soldan sağa doğru sırayla gezer.


##**Binary Tree'nin Kullanım Alanları**

Hiyerarşik veri modelleri: Dosya sistemleri, organizasyon yapıları vb.

Sıralama ve Arama Algoritmaları: Binary Search Tree (BST) kullanılarak hızlı aramalar yapılır.

Grafiksel Veri Modelleri: Özellikle bilgisayar grafikleri ve oyun programlama.

Özyineleme Tabanlı Sorunlar: Problem çözümlerinde özyinelemeli algoritmalar için ideal veri yapısıdır.

```python
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

# Binary Tree oluşturma
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
```

Bu basit yapıda bir Binary Tree oluşturuyoruz:

Kök düğüm 1 değerine sahip. Kökün solunda 2 ve sağında 3 bulunuyor. Sol çocuk olan 2'nin solunda 4, sağında ise 5 var. Şimdi ağaç yapısındaki bu yapıyı görselleştirelim:

```python
        1
       / \
      2   3
     / \
    4   5
```

```python
#Bu fonksiyonlar farklı sıralamalara göre ağaç üzerinde gezinti yapmanı sağlar.

# Pre-order traversal: Kök → Sol → Sağ
def pre_order_traversal(root):
    if root:
        print(root.val, end=" ")
        pre_order_traversal(root.left)
        pre_order_traversal(root.right)

# In-order traversal: Sol → Kök → Sağ
def in_order_traversal(root):
    if root:
        in_order_traversal(root.left)
        print(root.val, end=" ")
        in_order_traversal(root.right)

# Post-order traversal: Sol → Sağ → Kök
def post_order_traversal(root):
    if root:
        post_order_traversal(root.left)
        post_order_traversal(root.right)
        print(root.val, end=" ")
```

Traversal (gezinme), bir ağacın düğümlerini belirli bir sırada ziyaret etme işlemidir. Binary Tree'de en yaygın dört traversal yöntemi vardır:



##**1. Pre-order Traversal (Önce Kök)**

Pre-order traversal’da, düğümleri şu sıraya göre gezeriz:

Kök (root) düğümünü ziyaret et.

Sol alt ağacı (left subtree) ziyaret et.

Sağ alt ağacı (right subtree) ziyaret et.

**Adımlar:**

Önce kök düğüm (1) ziyaret edilir. Daha sonra sol çocuk (2) ziyaret edilir, ardından onun sol çocuğu (4) ve sağ çocuğu (5). En son sağ çocuk düğüm (3) ziyaret edilir.

**Pre-order Traversal Sırası: 1 → 2 → 4 → 5 → 3**

```python
def pre_order_traversal(root):
    if root:
        print(root.val, end=" ")
        pre_order_traversal(root.left)
        pre_order_traversal(root.right)

pre_order_traversal(root)
```

##**2. In-order Traversal (Sıralı)**

In-order traversal’da, şu sıraya göre düğümleri ziyaret ederiz:

Sol alt ağacı ziyaret et.

Kök düğümünü ziyaret et.

Sağ alt ağacı ziyaret et.

**Adımlar:**

Önce sol alt ağaç, yani 2 ve onun solundaki 4 ziyaret edilir.
Daha sonra kök düğüm (1) ziyaret edilir. Son olarak sağ alt ağaç, yani 3 ziyaret edilir.

**In-order Traversal Sırası: 4 → 2 → 5 → 1 → 3**

```python
def in_order_traversal(root):
    if root:
        in_order_traversal(root.left)
        print(root.val, end=" ")
        in_order_traversal(root.right)

in_order_traversal(root)
```

##**3. Post-order Traversal (Sonra Kök)**

Post-order traversal’da ise şu sıra izlenir:

Sol alt ağacı ziyaret et.

Sağ alt ağacı ziyaret et.

Kök düğümünü ziyaret et.

**Adımlar:**

Önce sol alt ağaç (2, 4, 5) ziyaret edilir. Sonra sağ alt ağaç (3) ziyaret edilir. En sonunda kök düğüm (1) ziyaret edilir.

**Post-order Traversal Sırası: 4 → 5 → 2 → 3 → 1**

```python
def post_order_traversal(root):
    if root:
        post_order_traversal(root.left)
        post_order_traversal(root.right)
        print(root.val, end=" ")

post_order_traversal(root)
```

##**4. Level-order Traversal (Seviye Seviyesi Gezinme)**

Level-order traversal, düğümleri ağacın her seviyesini soldan sağa doğru ziyaret eder.

**Adımlar:**

İlk olarak kök düğüm (1), ardından onun altındaki sol (2) ve sağ (3) düğümler ziyaret edilir. Sonra en alt seviyedeki düğümler, yani 4 ve 5, sırasıyla ziyaret edilir.

**Level-order Traversal Sırası: 1 → 2 → 3 → 4 → 5**

```python
from collections import deque

def level_order_traversal(root):
    if not root:
        return
    queue = deque([root])
    while queue:
        node = queue.popleft()
        print(node.val, end=" ")
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)

level_order_traversal(root)
```

Bu traversal yöntemlerini anladığımızda, her birinin kullanım amacını ve etkilerini de göz önünde bulundurabiliriz. In-order traversal, özellikle binary search tree (BST) yapısında sıralı veri elde etmek için kullanılırken; Post-order traversal, genellikle düğümlerin silinmesinde veya ağacın yapısının analiz edilmesinde kullanılır.

###Örneğin:
In-order traversal: Sıralama işlemleri, binary search tree'lerde arama işlemlerinde,

Pre-order traversal: Ağaç yapısının bir kopyasını oluşturma, matematiksel ifadelerin infix'ten prefix'e dönüştürme işlemlerinde,

Post-order traversal: Bir ifade ağacındaki ifadeyi değerlendirme, bir dizinin tüm alt dizilerini bulma işlemlerinde,

Level-order traversal: Bir ağacın genişliğini bulma, level-order tree traversal algoritmalarında kullanılır.

# **BINARY SEARCH TREE**

Binary Search Tree (BST), verilerin düzenli bir şekilde saklandığı ve hızlı bir şekilde erişildiği bir ağaç veri yapısıdır. BST’nin temel özellikleri şunlardır:

##**Binary Search Tree'nin Temel Özellikleri**

Her düğümün sol alt ağacı, o düğümün anahtar değerinden küçük değerlere sahip düğümler içerir.

Her düğümün sağ alt ağacı, o düğümün anahtar değerinden büyük değerlere sahip düğümler içerir.

Her düğümün alt ağaçları da kendi başlarına birer Binary Search Tree’dir.
Düğümler arasında tekrarlı değerler olmaz.

Bu özellikler, BST'nin veri arama, ekleme ve silme işlemlerinde O(log n) ortalama zaman karmaşıklığına sahip olmasını sağlar. Bu, veri yapılarını çok hızlı bir şekilde işleyebildiği anlamına gelir.

##**Binary Search Tree İşlemleri**

###**1. Arama (Search)**
Binary Search Tree'de arama, kök düğümden başlar ve aranan değer ile karşılaştırmalar yapılır:

Aranan değer kök düğümdeyse işlem biter.

Aranan değer kök düğümden küçükse sol alt ağaçta arama yapılır.

Aranan değer kök düğümden büyükse sağ alt ağaçta arama yapılır.

###**2. Ekleme (Insertion)**

Yeni bir düğüm eklenirken ağaç kök düğümden itibaren gezilerek uygun konum bulunur.

Yeni düğüm, uygun sol ya da sağ alt ağaçta olması gereken yere yerleştirilir.

###**3. Silme (Deletion)**

Bir düğüm silinirken üç olası durum vardır:

1. Yaprak düğüm (Leaf Node): Düğüm doğrudan silinir.

2. Tek çocuklu düğüm: Düğüm, çocuğu ile değiştirilir.

3. İki çocuklu düğüm: Sağ alt ağacın en küçük elemanı ya da sol alt ağacın en büyük elemanı bulunur ve silinmek istenen düğüm bu elemanla değiştirilir.

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    # Arama fonksiyonu
    def search(self, key):
        return self._search_recursive(self.root, key)

    def _search_recursive(self, node, key):
        # Düğüm bulunamadı
        if node is None or node.key == key:
            return node

        # Aranan değer küçükse, sol alt ağaca git
        if key < node.key:
            return self._search_recursive(node.left, key)

        # Aranan değer büyükse, sağ alt ağaca git
        return self._search_recursive(node.right, key)

    # Ekleme fonksiyonu
    def insert(self, key):
        if self.root is None:
            self.root = Node(key)
        else:
            self._insert_recursive(self.root, key)

    def _insert_recursive(self, node, key):
        if key < node.key:
            if node.left is None:
                node.left = Node(key)
            else:
                self._insert_recursive(node.left, key)
        elif key > node.key:
            if node.right is None:
                node.right = Node(key)
            else:
                self._insert_recursive(node.right, key)

    # Silme fonksiyonu
    def delete(self, key):
        self.root = self._delete_recursive(self.root, key)

    def _delete_recursive(self, node, key):
        if node is None:
            return node

        if key < node.key:
            node.left = self._delete_recursive(node.left, key)
        elif key > node.key:
            node.right = self._delete_recursive(node.right, key)
        else:
            # Tek veya hiç çocuğu olmayan düğüm
            if node.left is None:
                return node.right
            elif node.right is None:
                return node.left

            # İki çocuğu olan düğüm: Sağ alt ağacın en küçük elemanı ile değiştir
            min_larger_node = self._find_min(node.right)
            node.key = min_larger_node.key
            node.right = self._delete_recursive(node.right, min_larger_node.key)

        return node

    # Minimum değeri bulma fonksiyonu
    def _find_min(self, node):
        current = node
        while current.left is not None:
            current = current.left
        return current

    # Ağaç içeriğini sıralı yazdırma (Inorder Traversal)
    def inorder(self):
        self._inorder_recursive(self.root)

    def _inorder_recursive(self, node):
        if node:
            self._inorder_recursive(node.left)
            print(node.key, end=' ')
            self._inorder_recursive(node.right)

# Örnek kullanım
if __name__ == "__main__":
    bst = BinarySearchTree()

    # Eleman ekleme
    bst.insert(50)
    bst.insert(30)
    bst.insert(20)
    bst.insert(40)
    bst.insert(70)
    bst.insert(60)
    bst.insert(80)

    # Ağaç içeriğini yazdırma
    print("Inorder Traversal: ", end='')
    bst.inorder()
    print()

    # Arama
    if bst.search(40):
        print("40 bulundu")
    else:
        print("40 bulunamadı")

    # Silme
    bst.delete(20)
    print("20 silindi")

    # Ağaç içeriğini tekrar yazdırma
    print("Inorder Traversal after deletion: ", end='')
    bst.inorder()
    print()
```

###Kod için açıklamalar

**Düğüm Sınıfı (Node):**
Her bir düğüm, bir anahtar değeri (key), sol alt düğüm (left) ve sağ alt düğüm (right) referansları içerir.

**BinarySearchTree Sınıfı:** Bu sınıf, BST üzerinde arama, ekleme, silme ve inorder traversal işlemlerini içerir.

**Arama (search):** Bu yöntem, BST'de belirli bir değeri arar ve düğüm bulunursa onu döner.

**Ekleme (insert):** Yeni bir düğüm eklemek için kullanılır.

**Silme (delete):** Belirtilen anahtar değerine sahip düğümü siler.

**Sıralı Yazdırma (inorder):** Ağacı sıralı bir şekilde (sol, kök, sağ) yazdırır.

#**Zaman ve Alan Karmaşıklıkları**

Arama, Ekleme ve Silme işlemlerinin zaman karmaşıklığı, ağacın dengesine bağlı olarak $O(h)$'dir. Bu durumda $h$, ağacın yüksekliğidir.

Dengeli bir ağaçta bu karmaşıklık $O(log n)$ olur.

Dengesiz bir ağaçta (örneğin tamamen bir doğrusal yapıda) en kötü durumda $O(n)$ olabilir.

Binary Search Tree, verinin düzgün bir şekilde düzenlenmesi durumunda hızlı bir şekilde arama, ekleme ve silme işlemleri sağlar. Ancak ağaç dengesiz hale gelirse bu işlemler yavaşlayabilir. Bu sorunu çözmek için AVL ağacı ve Red-Black ağacı gibi dengeli ikili arama ağaçları tercih edilebilir.

# **AVL TREES**

Dengesiz ağaçları dengelemek için kullanılan en yaygın yöntemlerden biri AVL (Adelson-Velsky and Landis) ağacı'dır. AVL ağacı, her düğümde sol ve sağ alt ağaçların yükseklik farkını sürekli izleyerek dengeyi sağlar. Bu dengeleme işlemi, rotasyonlar adı verilen işlemlerle yapılır.

##**Dengesiz Ağaç Nedir?**

Bir dengesiz ikili arama ağacı, dallar arasındaki yükseklik farkının çok büyüdüğü, yani bir alt ağacın diğerine göre çok daha uzun olduğu durumlardır. Bu, performansı olumsuz etkileyebilir, çünkü arama, ekleme ve silme işlemlerinde ağaç yapısı dengeli olmadığında $O(n)$ zaman karmaşıklığına neden olabilir. İdeal durumda, bir ağacın dengeli olması için, her düğümde sol ve sağ alt ağaçların yüksekliği arasındaki fark 1'den büyük olmamalıdır.

##**AVL Ağacı Nedir?**

AVL ağacı, her düğümde sol ve sağ alt ağaçların yükseklik farkının en fazla 1 olmasını garanti eden bir dengeli ikili arama ağacı türüdür. Her düğümdeki bu fark, düğümün denge faktörü (balance factor) olarak adlandırılır ve şu şekilde hesaplanır:

Denge Faktörü = Sol alt ağacın yüksekliği-Sağ alt ağacın yüksekliği

Eğer bir düğümdeki denge faktörü -1, 0 veya 1 değilse, ağaç dengesiz demektir ve rotasyon adı verilen işlemlerle bu denge sağlanmalıdır.

##**AVL Ağacında Rotasyonlar**

AVL ağacında dengesizlik oluştuğunda, dört farklı durum için rotasyon uygulanır. Bu rotasyonlar şu şekildedir;

###**1. LL (Left-Left) Rotasyonu:**
Eğer bir düğümün sol alt ağacı ağırsa (yani, denge faktörü 2) ve sol çocuğun sol alt ağacı da ağırsa, bu durumda LL rotasyonu yapılır.

Sol alt ağaç tek taraflı büyüdüğünde, sağa doğru tek rotasyon yapılarak denge sağlanır.

###**2. RR (Right-Right) Rotasyonu:**
Eğer bir düğümün sağ alt ağacı ağırsa (denge faktörü -2) ve sağ çocuğun sağ alt ağacı da ağırsa, bu durumda RR rotasyonu yapılır.

Sağ alt ağaç tek taraflı büyüdüğünde, sola doğru tek rotasyon yapılır.

###**3. LR (Left-Right) Rotasyonu:**
Eğer sol alt ağaç ağırsa (denge faktörü 2) fakat sol çocuğun sağ alt ağacı ağırsa, bu durumda LR rotasyonu yapılır.

Sol-alt-sağ büyümesi olduğunda, önce sola küçük bir rotasyon yapılır (LL), ardından sağa bir rotasyon yapılır.

###**4. RL (Right-Left) Rotasyonu:**
Eğer sağ alt ağaç ağırsa (denge faktörü -2) fakat sağ çocuğun sol alt ağacı ağırsa, bu durumda RL rotasyonu yapılır.

Sağ-alt-sol büyümesi olduğunda, önce sağa küçük bir rotasyon yapılır (RR), ardından sola bir rotasyon yapılır.

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None
        self.height = 1  # Yüksekliği takip etmek için her düğümde 'height' tutulur

class AVLTree:
    def insert(self, root, key):
        # Normal BST ekleme işlemi
        if not root:
            return Node(key)
        elif key < root.key:
            root.left = self.insert(root.left, key)
        else:
            root.right = self.insert(root.right, key)

        # Yüksekliği güncelle
        root.height = 1 + max(self.get_height(root.left), self.get_height(root.right))

        # Denge faktörünü hesapla
        balance = self.get_balance(root)

        # Dengesizlik durumları
        # LL Durumu
        if balance > 1 and key < root.left.key:
            return self.right_rotate(root)

        # RR Durumu
        if balance < -1 and key > root.right.key:
            return self.left_rotate(root)

        # LR Durumu
        if balance > 1 and key > root.left.key:
            root.left = self.left_rotate(root.left)
            return self.right_rotate(root)

        # RL Durumu
        if balance < -1 and key < root.right.key:
            root.right = self.right_rotate(root.right)
            return self.left_rotate(root)

        return root

    # Sol rotasyon işlemi
    def left_rotate(self, z):
        y = z.right
        T2 = y.left

        # Rotasyon işlemi
        y.left = z
        z.right = T2

        # Yükseklikleri güncelle
        z.height = 1 + max(self.get_height(z.left), self.get_height(z.right))
        y.height = 1 + max(self.get_height(y.left), self.get_height(y.right))

        return y

    # Sağ rotasyon işlemi
    def right_rotate(self, z):
        y = z.left
        T3 = y.right

        # Rotasyon işlemi
        y.right = z
        z.left = T3

        # Yükseklikleri güncelle
        z.height = 1 + max(self.get_height(z.left), self.get_height(z.right))
        y.height = 1 + max(self.get_height(y.left), self.get_height(y.right))

        return y

    # Yükseklik hesaplama
    def get_height(self, node):
        if not node:
            return 0
        return node.height

    # Denge faktörünü hesaplama
    def get_balance(self, node):
        if not node:
            return 0
        return self.get_height(node.left) - self.get_height(node.right)

    # Inorder traversal (ağaç içeriğini yazdırma)
    def inorder(self, root):
        if not root:
            return
        self.inorder(root.left)
        print(f"{root.key} ", end="")
        self.inorder(root.right)

# AVL ağacını test etme
if __name__ == "__main__":
    tree = AVLTree()
    root = None

    # Elemanları ekleme
    numbers = [10, 20, 30, 40, 50, 25]
    for number in numbers:
        root = tree.insert(root, number)

    # Ağaç içeriğini yazdırma
    print("AVL Tree'nin inorder traversal yazılımı: ", end="")
    tree.inorder(root)
    print()
```

##Kod için açıklamalar,

**Düğüm Sınıfı (Node):**
Her düğümün, anahtar değeri (key), sol ve sağ çocukları (left, right), ve yüksekliği (height) bulunur.

**AVL Ağaç Sınıfı:**
AVL ağacı işlemlerini içerir.

**insert:** AVL ağacına eleman ekler ve gerekirse rotasyonları uygular.

**left_rotate ve right_rotate:** Dengesizlik durumlarında rotasyonları gerçekleştirir.

**get_height:** Bir düğümün yüksekliğini döner.

**get_balance:** Bir düğümün denge faktörünü hesaplar.

**inorder:** Ağacı sıralı bir şekilde yazdırır.

##**Zaman Karmaşıklığı**

AVL ağacında arama, ekleme, silme işlemleri, **O(log (n))** zaman karmaşıklığına sahiptir. Çünkü AVL ağacında denge her zaman korunur.

AVL ağacı, performansı korumak için ağaç dengesini sağlamak amacıyla rotasyon işlemleri kullanarak dengesiz ağaç problemini çözer.

##**NERELERDE KULLANILIR**

# **HEAP TREE**

Heap Tree, genellikle öncelik sıralı işlemler için kullanılan bir ağaç veri yapısıdır ve iki ana türü vardır: Max-Heap ve Min-Heap.

Heap'ler genellikle binary tree'lerden (ikili ağaç) biri olan complete binary tree (tam ikili ağaç) olarak uygulanırlar. Bu, her seviyenin dolu olması gerektiği anlamına gelir, yalnızca son seviyede boşluk olabilir ve bu boşluklar en sağda olmalıdır.

Heap Tree’nin özelliklerini ve uygulanmasını inceleyelim.

##**1. Heap'in Temel Özellikleri**

###**a. Max-Heap**
Max-Heap'te her bir düğüm, alt düğümlerinden büyük veya eşittir.
Kök düğümü (root) her zaman en büyük elemandır.

###**b. Min-Heap**

Min-Heap'te her düğüm, alt düğümlerinden küçük veya eşittir.
Kök düğümü her zaman en küçük elemandır.

###**c. Complete Binary Tree**

Heap, eksiksiz bir ikili ağaçtır; her seviyedeki düğümler mümkün olduğunca doldurulmalıdır.

##**2. Heap Tree İşlemleri**

Heap, bir dizi işlemi verimli bir şekilde gerçekleştirebilir, bu nedenle veri sıralama, öncelik kuyruğu ve benzeri algoritmalarda sıkça kullanılır.


###**a. Insert (Ekleme)**

Yeni bir düğüm eklerken ağacın yapısını bozmamak için son pozisyona eklenir.
Sonrasında, yukarı hareket ederek (heapify-up) heap'in özelliklerini sağlamak için yeni düğüm diğer düğümlerle karşılaştırılır.

**Zaman karmaşıklığı: $O(log (n))$**

###**b. Delete (Silme)**

Kök düğümü (en büyük ya da en küçük eleman) silinir.

Son yaprak, köke taşınır.

Sonrasında, aşağı hareket ederek (heapify-down) heap'in özellikleri korunur.

**Zaman karmaşıklığı: $O(log (n))$**

###**c. Peek (Göz Atma)**

Kök düğümüne bakmak (en büyük veya en küçük eleman) işlemi $O(1)$ zaman alır.

##**3. Heap Sort (Heap Sıralama)**

Heap kullanılarak veri sıralaması yapılabilir. Bu yöntem Heap Sort olarak adlandırılır ve şu adımları içerir:

Max-Heap yapısı kurulur.

Kök düğümü (en büyük eleman) alınır ve dizinin sonuna konur.

Yeni kök düğümü olarak son eleman alınır ve yeniden heapify-down işlemi uygulanarak heap yeniden yapılandırılır.

Bu işlem, sıralanmış bir dizi elde edene kadar devam eder.

##**4. Heap'in Array ile Temsili**

Heap'ler, verimli bir şekilde array (dizi) ile temsil edilebilir. Bu temsilde:

Kök düğümü dizi başında bulunur: A[0]

Bir düğümün sol çocuğu $2 * i + 1$, sağ çocuğu ise $2 * i + 2$ indeksinde bulunur.

Bir düğümün ebeveyni ise $(i - 1) / 2$ indeksindedir.







```python
class MaxHeap:
    def __init__(self):
        self.heap = []

    def parent(self, i):
        return (i - 1) // 2

    def leftChild(self, i):
        return 2 * i + 1

    def rightChild(self, i):
        return 2 * i + 2

    def insert(self, key):
        self.heap.append(key)
        self._heapify_up(len(self.heap) - 1)

    def delete(self):
        if len(self.heap) == 0:
            return None
        if len(self.heap) == 1:
            return self.heap.pop()

        root = self.heap[0]
        self.heap[0] = self.heap.pop()  # Son elemanı köke taşıyoruz
        self._heapify_down(0)  # Heap'i aşağı doğru yeniden düzenliyoruz
        return root

    def _heapify_up(self, index):
        # Yukarı doğru hareket ederek heap'in dengesini sağlıyoruz
        while index != 0 and self.heap[self.parent(index)] < self.heap[index]:
            self.heap[self.parent(index)], self.heap[index] = self.heap[index], self.heap[self.parent(index)]
            index = self.parent(index)

    def _heapify_down(self, index):
        # Aşağı doğru hareket ederek heap'in dengesini sağlıyoruz
        largest = index
        left = self.leftChild(index)
        right = self.rightChild(index)

        if left < len(self.heap) and self.heap[left] > self.heap[largest]:
            largest = left

        if right < len(self.heap) and self.heap[right] > self.heap[largest]:
            largest = right

        if largest != index:
            self.heap[index], self.heap[largest] = self.heap[largest], self.heap[index]
            self._heapify_down(largest)

    def get_max(self):
        if len(self.heap) == 0:
            return None
        return self.heap[0]

# Kullanım:
heap = MaxHeap()
heap.insert(10)
heap.insert(20)
heap.insert(15)
heap.insert(30)
print("Max Eleman:", heap.get_max())  # Max Eleman: 30
print("Kök Eleman Silindi:", heap.delete())  # Kök Eleman Silindi: 30
print("Yeni Max Eleman:", heap.get_max())  # Yeni Max Eleman: 20
```

##**5. Heap'in Avantajları ve Dezavantajları**

###**Avantajları**

Hızlı en büyük/en küçük eleman bulma: Zaman karmaşıklığı $O(1)$ olduğundan heap'ler hızlıdır.

Öncelik Kuyruğu: Dinamik veri yapılarında, en büyük ya da en küçük elemana sürekli olarak hızlı bir şekilde erişilebilir.

Heap Sort: Sıralama algoritmaları arasında en verimli algoritmalardan biridir.

###**Dezavantajları**

Verimsiz arama işlemleri: Belirli bir değeri bulmak $O(n)$ zaman alabilir.

Dengesiz Ağaçlar: Heap'ler bir binary search tree olmadığından, dengesiz ağaç yapısına sahip olabilirler. Her düğümün iki çocuğu olmasa da heap'in işlevi etkilenmez.


Heap Tree, özellikle verilerin hızlı bir şekilde sıralanması ve öncelik sırasına göre işlenmesi gereken durumlarda oldukça etkilidir. Min-Heap ve Max-Heap türlerine sahip olan bu yapı, sıklıkla veri tabanı işlemleri, iş zamanlayıcıları ve öncelik kuyruğu uygulamalarında kullanılır.

# **TRIES**

Trie (pronounced as "try"), özellikle string verilerini verimli bir şekilde saklamak ve aramak için kullanılan bir ağaç (tree) yapısıdır.Trie ağacının ismi retrieval kelimesininin ortasındaki 4 harften gelmektedir. Trie'ler, genellikle kelimelerin bir kümesini saklamak için kullanılır ve prefix aramaları (önek aramaları) için idealdir.

1. **Trie Nedir?**

Trie, bir veri yapısıdır ve her düğüm bir karakteri temsil eder. Bu yapının en yaygın kullanım alanı kelime listeleri veya sözlükler ile ilgilidir. Örneğin, bir Trie yapısında kelimelerin her karakteri, bir düğümde saklanır ve bir kelimenin tamamı bir yol (path) oluşturur.

2. **Trie Yapısının Özellikleri**

Her düğüm bir karakteri temsil eder: Her düğüm bir karakter içerir ve bir kelimenin tüm karakterlerini saklamak için bir yol oluşturulur.

Kök düğüm boş: Kök düğüm (root), herhangi bir karakter içermez.

Kelimenin sonunu işaretlemek: Trie'de bir kelimenin sonuna gelindiğinde, genellikle bir işaretçi kullanılarak bu düğümde bir kelimenin bittiği gösterilir.

###**Trie'nin Avantajları**

Trie'nin kullanıldığı başlıca durumlar şunlardır:

Hızlı Prefix Araması: Trie'ler, belirli bir önekle başlayan tüm kelimeleri hızlıca bulmanızı sağlar. Örneğin, bir kelime sözlüğünde "app" ile başlayan tüm kelimeleri bulmak için etkili bir yapıdır.

Hızlı Arama: Trie, kelimelerin doğruluğunu veya var olup olmadığını hızlıca kontrol etmenizi sağlar. Örneğin, bir kelimenin sözlükte olup olmadığını O(m) zamanda (m, kelimenin uzunluğu) kontrol edebilirsiniz.

###**Trie'nin Dezavantajları**

Yer Kullanımı: Trie'ler, her karakter için bir düğüm oluşturduğu için büyük miktarda hafıza kullanabilir. Eğer bir veri setinde çok fazla kelime varsa, hafıza kullanımı yüksek olabilir.

3. **Trie'nin Yapısı ve Temel İşlemleri**

Trie veri yapısında yapılan üç temel işlem vardır: ekleme (insert), arama (search), silme (delete).

**a. Ekleme (Insert)**

Bir kelimeyi Trie'ye eklerken, kökten başlarız ve her karakteri bir düğüm olarak ekleriz. Eğer bir karakter zaten varsa, mevcut düğümle devam ederiz.

**b. Arama (Search)**

Bir kelimeyi Trie'de ararken, kelimenin her karakterini kökten başlayarak ararız. Eğer karakterlerin tamamını bulursak ve kelimenin sonunu işaret eden bir düğüm varsa, kelime Trie'de vardır.

**c. Silme (Delete)**

Bir kelimeyi silerken, önce kelimenin Trie'de olup olmadığını kontrol ederiz. Eğer kelime Trie'de varsa, kelimenin düğümlerini silmeye başlarız. Eğer düğümün başka bir kelimenin parçası olmadığı belirlenirse, bu düğüm ve alt düğümleri silinir.

```python
class TrieNode:
    def __init__(self):
        self.children = {}  # Çocuk düğümleri depolamak için bir sözlük kullanılır
        self.is_end_of_word = False  # Bu düğümün bir kelimenin sonu olup olmadığını gösterir

class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        """
        Trie'ye bir kelime ekler.
        """
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()  # Eğer karakter yoksa, yeni bir düğüm oluştur
            node = node.children[char]
        node.is_end_of_word = True  # Kelimenin sonuna gelindiğinde işaretlenir
        print(f"{word}")

    def search(self, word):
        """
        Trie'de bir kelime arar. Kelime varsa True, yoksa False döner.
        """
        node = self.root
        for char in word:
            if char not in node.children:
                return False  # Eğer karakter yoksa, kelime de yoktur
            node = node.children[char]
        return node.is_end_of_word  # Son düğümde kelimenin sonu olup olmadığına bakılır


    def starts_with(self, prefix):
        """
        Verilen önekle başlayan kelimelerin olup olmadığını kontrol eder.
        """
        node = self.root
        for char in prefix:
            if char not in node.children:
                return False  # Önek bulunamazsa False döner
            node = node.children[char]
        return True  # Eğer önek bulunursa True döner

    def delete(self, word):
        """
        Trie'den bir kelimeyi siler.
        """
        def _delete(node, word, depth):
            # Tüm kelime işlendiyse ve bu bir kelimenin sonuysa, işaretlemeyi kaldır
            if depth == len(word):
                if not node.is_end_of_word:
                    return False  # Kelime Trie'de bulunmuyorsa
                node.is_end_of_word = False
                return len(node.children) == 0  # Eğer yaprak düğümse silmek için True döner

            # Rekürsif olarak düğümleri silmek için çağrı
            char = word[depth]
            if char not in node.children:
                return False  # Eğer kelime Trie'de bulunmuyorsa

            should_delete_current_node = _delete(node.children[char], word, depth + 1)

            # Çocuk düğümü sil
            if should_delete_current_node:
                del node.children[char]
                return len(node.children) == 0

            return False

        _delete(self.root, word, 0)

# Kullanım:
trie = Trie()
trie.insert("apple")
print(trie.search("apple"))  # True
print(trie.search("app"))  # False
print(trie.starts_with("app"))  # True
trie.insert("app")
print(trie.search("app"))  # True
trie.delete("apple")
print(trie.search("apple"))  # False
```

# **FENWICK TREES**

Fenwick Tree, bir dizi üzerinde aralık sorguları ve güncellemeler yapmak için kullanılan, özellikle toplama işlemleri için optimize edilmiş bir veri yapısıdır. Bir Fenwick Tree, toplama sorgularını ve güncellemeleri
$𝑂(log𝑛)$ sürede yapabilir. Bu veri yapısı, Segment Tree'ye göre daha az yer kaplar ve bazı durumlarda daha hızlı olabilir.

##**Fenwick Tree'nin temel özellikleri**

**Aralık Toplamı:**
Dizinin belirli bir alt aralığındaki (örneğin [1, k] aralığındaki) elemanların toplamını verimli bir şekilde hesaplar.

**Güncelleme:** Bir elemanın değeri değiştirildiğinde, diziyi verimli bir şekilde günceller.

Fenwick Tree, indekslerin ikili gösteriminde kullanılan belirli bitlere dayanarak aralıkları bölerek işlemlerini gerçekleştirir. Bu sayede çok hızlı sonuçlar alınabilir.

##**Fenwick Tree Teorisi**

Fenwick Tree, bir dizinin elemanlarına dayalı bir yapıdır ve dizinin her indeksine bir ikili sayı gözüyle bakılır. Dizideki her bir elemanın toplamı, en sağdaki 1'den sonraki tüm bitleri kapsayan alt aralığın toplamını temsil eder.

Örneğin, bir dizinin 6. elemanını ele alalım (indeksleme 1'den başlar):

6 = 110₂, bu durumda 6'daki toplam, [5, 6] aralığını kapsar.

5 = 101₂, bu durumda 5'teki toplam, [5, 5] aralığını kapsar.

Her güncelleme işleminde, ilgili indeksin üst aralıklarına eklemeler yapılır. Benzer şekilde, sorgulama yaparken daha geniş alt aralıkların toplamlara dahil edilmesi için indeksler azaltılarak ilerlenir.

##**Fenwick Tree'nin Aşamaları**

**Sorgu:** [1, k] aralığındaki elemanların toplamını bulma.

**Güncelleme:** Dizideki bir değeri artırmak ya da azaltmak.
Bu işlemleri $𝑂(log (𝑛))$ sürede yapabiliyor olmamız, Fenwick Tree'nin avantajıdır.

**Python Koduyla Fenwick Tree (BIT) Uygulaması**

Aşağıda Fenwick Tree'yi (Binary Indexed Tree) Python'da nasıl uygulayabileceğimizi adım adım açıklayan bir kod örneği bulunmaktadır:

```python
class FenwickTree:
    def __init__(self, n):
        """Fenwick Tree oluşturma. Boyut n olan bir diziyle çalışacağız."""
        self.n = n
        self.tree = [0] * (n + 1)  # 1'den indekslenen dizimiz

    def update(self, index, value):
        """index numaralı elemanı value kadar güncelle."""
        while index <= self.n:
            self.tree[index] += value
            index += index & -index  # Sağdaki 1'i bul ve ekle

    def query(self, index):
        """1'den index'e kadar olan toplamı döndür."""
        result = 0
        while index > 0:
            result += self.tree[index]
            index -= index & -index  # Sağdaki 1'i çıkar ve bir önceki alt gruba geç
        return result

    def range_query(self, left, right):
        """left ile right arasındaki toplamı döndür."""
        return self.query(right) - self.query(left - 1)

# Örnek kullanım
if __name__ == "__main__":
    # 8 elemanlı bir dizimiz var.
    ft = FenwickTree(8)

    # Dizimize bazı değerler ekleyelim.
    values = [0, 3, 2, -1, 6, 5, 4, -3, 3]  # 1'den indekslenmiş
    for i in range(1, len(values)):
        ft.update(i, values[i])

    # Sorgulamalar yapalım
    print("1'den 5'e kadar olan toplam:", ft.query(5))  # 1'den 5'e kadar olan toplam
    print("3 ile 7 arasındaki toplam:", ft.range_query(3, 7))  # 3'ten 7'ye kadar olan toplam
```

**Kodun Açıklaması**

**__init__ metodu:** Fenwick Tree'yi oluşturmak için kullanılır. Boyutu n olan bir dizi ile çalışır. Diziyi bir eleman fazlalığı ile başlatıyoruz (1'den indeksleniyor).

**update metodu:** Bir indeksteki değeri güncellemek için kullanılır. İndeksin ikili gösterimindeki en sağdaki 1 bitini kullanarak ilgili aralıkları güncelleriz.

**query metodu:** 1'den belirtilen indeks numarasına kadar olan toplamı döndürür. İndeks ikili gösteriminde sağdaki 1 bitini kullanarak toplama işlemi yapar.

**range_query metodu:** Belirtilen aralıktaki (left-right) toplamı döndürmek için kullanılır. Bunu iki query çağrısı yaparak hesaplar.

**Örnek**

Dizimiz: [3, 2, -1, 6, 5, 4, -3, 3]

Sorgu 1: ft.query(5):

elemandan 5. elemana kadar olan toplam 3+2+(−1)+6+5=15.

Sorgu 2: ft.range_query(3, 7):

3'ten 7'ye kadar olan toplam (−1)+6+5+4+(−3)=11.
