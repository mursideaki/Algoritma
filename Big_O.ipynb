Bu kurallarƒ±n dƒ±≈üƒ±nda, d√∂ng√º yapƒ±larƒ± ve i√ß i√ße d√∂ng√ºler gibi yapƒ±lar da Big O notasyonu ile analiz edilir. √ñrneƒüin, sabit sayƒ±da adƒ±m i√ßeren bir algoritma
ùëÇ(1), i√ß i√ße d√∂ng√ºler i√ßeren bir algoritma ise genellikle √ßarpanlarƒ±na g√∂re analiz edilip yazƒ±lƒ±r.

Verilen ifadeyi sadele≈ütirmek, Big O notasyonunun temel mantƒ±ƒüƒ±na dayanƒ±r: Algoritmanƒ±n √ßalƒ±≈üma s√ºresindeki en b√ºy√ºk b√ºy√ºme oranƒ±na odaklanmak ve k√º√ß√ºk terimlerle sabit katsayƒ±larƒ± g√∂z ardƒ± etmek.

Bu ifadeyi adƒ±m adƒ±m nasƒ±l sadele≈ütirdiƒüimizi anlamaya √ßalƒ±≈üalƒ±m:

Bu ifadede en b√ºy√ºk terim; $3n^5logn$ ifadesidir. √ústel bir ifade ve logaritma i√ßeren bu terim algoritmanƒ±n en hƒ±zlƒ± b√ºy√ºyecek olan kƒ±smƒ±dƒ±r. ƒ∞fadenin kalan kƒ±sƒ±mlarƒ±nƒ± inceleyelim $10ùëõ^4$ ifadesi de aslƒ±nda b√ºy√ºk bir sayƒ±yƒ± ifade edecektir ama $3n^5logn$'e g√∂re daha k√º√ß√ºkt√ºr, √ß√ºnk√º logaritmik b√ºy√ºme g√∂stermeyecektir. $ùëõ^2$ ise daha k√º√ß√ºk bir polinomdur $n^5$ daha hƒ±zlƒ± artacaktƒ±r. ùëõ ise lineer bir terimdir, daha k√º√ß√ºk bir b√ºy√ºme g√∂sterir. 15 sabit bir terimdir ve zaman karma≈üƒ±klƒ±ƒüƒ± a√ßƒ±sƒ±ndan en k√º√ß√ºk etkiye sahiptir.

**Farklƒ± Big O Notasyonlarƒ±**

**O(1) ‚Äì Sabit Zaman**

Tanƒ±m: Girdi boyutundan baƒüƒ±msƒ±z olarak sabit bir zaman alan algoritmalar.

√ñrnek: Bir dizinin ilk elemanƒ±nƒ± almak.

```python
def get_first_element(arr):
    return arr[0]
```

Burada dizinin boyutu ne olursa olsun ilk elemana eri≈ümek sabit zaman alƒ±r.

**O(n) ‚Äì Doƒürusal Zaman**

Tanƒ±m: Girdi boyutu n kadar arttƒ±ƒüƒ±nda i≈ülem sayƒ±sƒ± da n ile aynƒ± oranda artan algoritmalar.

√ñrnek: Bir dizideki t√ºm elemanlarƒ± toplamak.

```python
def sum_of_elements(arr):
    total = 0
    for num in arr:
        total += num
    return total
```

Her eleman i√ßin bir i≈ülem yapƒ±ldƒ±ƒüƒ± i√ßin zaman karma≈üƒ±klƒ±ƒüƒ± O(n)'dir.

**O(n¬≤) ‚Äì Kare Zaman**

Tanƒ±m: Girdinin boyutuna baƒülƒ± olarak, i≈ülem sayƒ±sƒ±nƒ±n n'nin karesi kadar arttƒ±ƒüƒ± algoritmalar.

√ñrnek: ƒ∞ki i√ß i√ße d√∂ng√º ile √ßalƒ±≈üan algoritmalar (√∂rneƒüin, Selection Sort).

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr
```

ƒ∞√ß i√ße d√∂ng√ºler nedeniyle bu algoritmanƒ±n zaman karma≈üƒ±klƒ±ƒüƒ± O(n¬≤)'dir.

**O(log n) ‚Äì Logaritmik Zaman**

Tanƒ±m: Girdi boyutu arttƒ±k√ßa, i≈ülem sayƒ±sƒ±nƒ±n logaritmik olarak arttƒ±ƒüƒ± algoritmalar.

√ñrnek: Binary Search.

```python
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

Her adƒ±mda arama alanƒ±nƒ± yarƒ±ya b√∂ld√ºƒü√º i√ßin O(log n) zaman karma≈üƒ±klƒ±ƒüƒ±na sahiptir.

**O(n log n) ‚Äì Lineer Logaritmik Zaman**

Tanƒ±m: √áoƒüu verimli sƒ±ralama algoritmasƒ±nƒ±n zaman karma≈üƒ±klƒ±ƒüƒ±dƒ±r.

√ñrnek: Merge Sort ve Quick Sort.

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1
    return arr
```

Merge Sort, her b√∂l√ºmleme i≈ülemi O(log n) ve birle≈ütirme i≈ülemi O(n) olduƒüundan O(n log n) karma≈üƒ±klƒ±ƒüƒ±na sahiptir.

**O(2‚Åø) ‚Äì √ústel Zaman**

Tanƒ±m: Girdi boyutu arttƒ±k√ßa i≈ülem sayƒ±sƒ±nƒ±n √ºstel olarak arttƒ±ƒüƒ± algoritmalar. Genellikle rek√ºrsif √ß√∂z√ºm y√∂ntemlerinde g√∂r√ºl√ºr.

√ñrnek: Rek√ºrsif y√∂ntemle Fibonacci dizisini hesaplamak.

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
```

Rek√ºrsif olarak √ß√∂z√ºm arandƒ±ƒüƒ± i√ßin zaman karma≈üƒ±klƒ±ƒüƒ± O(2‚Åø) olur

**O(n!) ‚Äì Fakt√∂riyel Zaman**

Tanƒ±m: √á√∂z√ºm uzayƒ±nƒ± tam taramak gereken durumlarda kullanƒ±lƒ±r.

√ñrnek: Perm√ºtasyon √ºretme algoritmalarƒ±.

```python
def generate_permutations(arr):
    if len(arr) == 0:
        return [[]]
    first = arr[0]
    rest = generate_permutations(arr[1:])
    permutations = []
    for perm in rest:
        for i in range(len(perm) + 1):
            permutations.append(perm[:i] + [first] + perm[i:])
    return permutations
```

T√ºm olasƒ± sƒ±ralamalarƒ± (perm√ºtasyonlarƒ±) bulduƒüu i√ßin fakt√∂riyel zaman karma≈üƒ±klƒ±ƒüƒ±na sahiptir.

**Asimptotik Notasyonlar**

Big O Notation, bir algoritmanƒ±n performansƒ±nƒ± 3 farklƒ± durumda analiz eder:


**Best-case:** En iyi senaryo / Omega notasyonu / Omega g√∂sterimi / Big-Œ©

**Worst-case:** En k√∂t√º senaryo / Big-O notasyonu / B√ºy√ºk O Notasyonu / Big O g√∂sterimi

##Big Omega (Œ©) Notasyonu

Anlamƒ±: Bir algoritmanƒ±n en iyi durumda (best case) ne kadar hƒ±zlƒ± √ßalƒ±≈üacaƒüƒ±nƒ±n alt sƒ±nƒ±rƒ±nƒ± verir. Ba≈üka bir deyi≈üle, algoritma asla bu s√ºreden daha kƒ±sa s√ºrede √ßalƒ±≈üamaz.

Matematiksel Tanƒ±m: $f(n) = Œ©(g(n))$ ise, pozitif sabitler c ve n‚ÇÄ i√ßin, t√ºm $n ‚â• n‚ÇÄ$ deƒüerleri i√ßin $0 ‚â§ c * g(n) ‚â§ f(n)$ e≈üitsizliƒüi saƒülanƒ±r.

Yani algoritmanƒ±n √ßalƒ±≈üma s√ºresi, $g(n)$ fonksiyonuyla orantƒ±lƒ± veya daha hƒ±zlƒ±dƒ±r.

√ñrnek: Bir sƒ±ralƒ± dizide bir eleman arama i≈ülemi i√ßin en iyi durum, aranƒ±lan elemanƒ±n dizinin ba≈üƒ±nda olmasƒ±dƒ±r. Bu durumda, sadece bir kar≈üƒ±la≈ütƒ±rma yapƒ±lacaƒüƒ± i√ßin zaman karma≈üƒ±klƒ±ƒüƒ± $Œ©(1)$'dir.

##Big Theta (Œò) Notasyonu

Anlamƒ±: Bir algoritmanƒ±n hem en iyi hem de en k√∂t√º durumdaki √ßalƒ±≈üma s√ºresini sƒ±kƒ± bir ≈üekilde sƒ±nƒ±rlar. Yani, algoritmanƒ±n √ßalƒ±≈üma s√ºresi her zaman belirli bir b√ºy√ºme oranƒ±na sahiptir.

Matematiksel Tanƒ±m: $f(n) = Œò(g(n))$ ise, pozitif sabitler $c‚ÇÅ, c‚ÇÇ$ ve $n‚ÇÄ$ i√ßin, t√ºm $n ‚â• n‚ÇÄ$ deƒüerleri i√ßin $0 ‚â§ c‚ÇÅ * g(n) ‚â§ f(n) ‚â§ c‚ÇÇ * g(n)$ e≈üitsizliƒüi saƒülanƒ±r.

Ne Anlama Gelir: Algoritmanƒ±n √ßalƒ±≈üma s√ºresi, $g(n)$ fonksiyonuyla aynƒ± b√ºy√ºme oranƒ±na sahiptir.

√ñrnek: Doƒürusal arama algoritmasƒ± i√ßin, en iyi ve en k√∂t√º durumlar arasƒ±nda √ßok b√ºy√ºk bir fark yoktur. Bu nedenle, doƒürusal arama algoritmasƒ±nƒ±n zaman karma≈üƒ±klƒ±ƒüƒ± $Œò(n)$ olarak ifade edilir.
