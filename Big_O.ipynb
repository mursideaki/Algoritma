# **Big O NOTASYONU**

Big O notasyonu, algoritmaların verimliliğini anlamak için kullanılır ve algoritmanın çalışması için gereken zamanı veya hafıza kullanımını, girdinin büyüklüğüne bağlı olarak tanımlar. Bu notasyon, özellikle en kötü durum senaryosuna odaklanır ve performansı değerlendirirken dikkate alınması gereken kritik bir kavramdır.

Temel Big O Kavramları
1. Zaman Karmaşıklığı (Time Complexity)

Zaman karmaşıklığı, bir algoritmanın ne kadar süre çalışacağını belirlemek için kullanılır. Ancak bu süreyi saniyeler cinsinden değil, algoritmanın gerçekleştirdiği işlem sayısına göre hesaplarız. Yapılan işlem sayısı, veri kümesinin büyüklüğüne (n) ve verilerin sıralamasına göre değişir. Zaman karmaşıklığını etkileyen faktörler arasında döngüler, karşılaştırmalar, rekürsif çağrılar ve diğer işlemler bulunur.

2. Alan Karmaşıklığı (Space Complexity)

Alan karmaşıklığı, bir algoritmanın hafızada ne kadar yer kapladığını ölçer. Bellek kullanımı, değişkenler, diziler, ek veri yapıları gibi faktörler tarafından belirlenir.



Big O Notasyonu, algoritmaların performansını analiz ederken kullanılan bir gösterim biçimidir. Bu notasyon, bir algoritmanın girdileri büyüdükçe zaman veya alan gereksinimlerinin nasıl değişeceğini öngörmemize yardımcı olur. Başka bir deyişle, algoritmanın ne kadar hızlı veya yavaş büyüdüğünü gösteren bir üst sınır belirtir.

Big O, algoritmaların karmaşıklığını sadeleştirerek ifade etmenin bir yoludur. Bu sayede, algoritmadaki küçük terimlerin veya sabitlerin zaman karmaşıklığı üzerindeki etkisi göz ardı edilerek büyük resme odaklanılır. Örneğin, bir algoritmanın zaman karmaşıklığı $3n^5logn+10n^4+n^2+n+15$ olarak verildiğinde, bu ifade sadeleştirilip $𝑂(3𝑛^5log𝑛)$ şeklinde yazılır. Yine benzer şekilde
$3n^2+3n+16$ gibi bir ifadeyi $𝑂(𝑛^2)$ olarak gösterebiliriz.

Big O notasyonu kullanırken bazı kurallar vardır:

**Katsayı Kuralı:** Katsayılar zaman karmaşıklığını etkilemez, bu yüzden
𝑓(𝑛)=𝑂(𝑔(𝑛)) olduğunda 𝑘𝑓(𝑛) yine 𝑂(𝑔(𝑛)) olur.

**Toplama Kuralı:** Farklı fonksiyonlar toplanırsa, daha baskın olan terim dikkate alınır.

**Çarpım Kuralı:** İki fonksiyon çarpıldığında karmaşıklıkları da çarpılır.

**Polinom Kuralı:** Polinomların en yüksek dereceli terimi kullanılır.

**Logaritmik Kuralı:** Logaritmik ifadelerde, logaritmanın derecesi zaman karmaşıklığını belirler.

Bu kuralların dışında, döngü yapıları ve iç içe döngüler gibi yapılar da Big O notasyonu ile analiz edilir. Örneğin, sabit sayıda adım içeren bir algoritma
𝑂(1), iç içe döngüler içeren bir algoritma ise genellikle çarpanlarına göre analiz edilip yazılır.

Burada anlattıklarımıza göre yukarıdaki örneği inceleyelim;

Verilen ifadeyi sadeleştirmek, Big O notasyonunun temel mantığına dayanır: Algoritmanın çalışma süresindeki en büyük büyüme oranına odaklanmak ve küçük terimlerle sabit katsayıları göz ardı etmek.

Verilen ifade şu şekilde: $3n^5logn+10n^4+n^2+n+15$

Bu ifadeyi adım adım nasıl sadeleştirdiğimizi anlamaya çalışalım:

Öncelikle terimlere bakıyoruz. Bu tür bir ifadede, algoritmanın en büyük büyüme hızına sahip olan terimine odaklanmalıyız. Çünkü Big O notasyonu, en yüksek dereceli terim ile algoritmanın nasıl büyüdüğünü belirtir çünkü diğer küçük terimler, giriş boyutu çok büyük olduğunda zaman karmaşıklığına pek fazla etki etmez. Bu ifadede en büyük terim; $3n^5logn$ ifadesidir. Üstel bir ifade ve logaritma içeren bu terim algoritmanın en hızlı büyüyecek olan kısmıdır. İfadenin kalan kısımlarını inceleyelim $10𝑛^4$ ifadesi de aslında büyük bir sayıyı ifade edecektir ama $3n^5logn$'e göre daha küçüktür, çünkü logaritmik büyüme göstermeyecektir. $𝑛^2$ ise daha küçük bir polinomdur $n^5$ daha hızlı artacaktır. 𝑛 ise lineer bir terimdir, daha küçük bir büyüme gösterir. 15 sabit bir terimdir ve zaman karmaşıklığı açısından en küçük etkiye sahiptir.

Big O notasyonu küçük terimleri gözardı edecek, sadece giriş boyutu çok büyük olan terimleri dikkate alacaktır. Bu durumda en baskın terimi korur ve daha küçük büyüme hızına sahip terimleri yok sayarız.

**Katsayıları yok saymak:** Big O notasyonu, sabit katsayıları dikkate almaz. Yani, terimdeki 3, 10 gibi katsayılar önemli değildir, sadece terimlerin büyüme hızına odaklanırız.

$3n^5logn$ terimindeki 3 sabiti önemsenmez, sadece $n^5logn$ kısmına odaklanırız

Sonuç olarak, en büyük terim olan $n^5logn$ ifadesi baskındır ve diğer terimler
𝑛 büyüdükçe göz ardı edilebilecek kadar küçük kalır. Bu yüzden ifade: $n^5logn$
şeklinde sadeleştirilir.

Özetle: Big O notasyonunda sadeleştirme, en hızlı büyüyen terime odaklanarak yapılır, daha küçük büyüme oranlarına sahip terimler ve sabitler göz ardı edilir. Bu şekilde algoritmanın performansı bağımsız şekilde değerlendirilmiş ve karmaşıklık sadeleştirilmiştir böylece algoritmaların karşılaştırılması kolaylaştırılmış olur.

Yine de bazı durumlarda sabitlerin de önemli olabileceğini unutmamak gerekir. Özellikle gömülü sistemler gibi kaynakların sınırlı olduğu durumlarda, sabitler de performansı etkileyebilir.

**Farklı Big O Notasyonları**

**O(1) – Sabit Zaman**

Tanım: Girdi boyutundan bağımsız olarak sabit bir zaman alan algoritmalar.

Örnek: Bir dizinin ilk elemanını almak.

```python
def get_first_element(arr):
    return arr[0]
```

Burada dizinin boyutu ne olursa olsun ilk elemana erişmek sabit zaman alır.

**O(n) – Doğrusal Zaman**

Tanım: Girdi boyutu n kadar arttığında işlem sayısı da n ile aynı oranda artan algoritmalar.

Örnek: Bir dizideki tüm elemanları toplamak.

```python
def sum_of_elements(arr):
    total = 0
    for num in arr:
        total += num
    return total
```

Her eleman için bir işlem yapıldığı için zaman karmaşıklığı O(n)'dir.

**O(n²) – Kare Zaman**

Tanım: Girdinin boyutuna bağlı olarak, işlem sayısının n'nin karesi kadar arttığı algoritmalar.

Örnek: İki iç içe döngü ile çalışan algoritmalar (örneğin, Selection Sort).

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr
```

İç içe döngüler nedeniyle bu algoritmanın zaman karmaşıklığı O(n²)'dir.

**O(log n) – Logaritmik Zaman**

Tanım: Girdi boyutu arttıkça, işlem sayısının logaritmik olarak arttığı algoritmalar.

Örnek: Binary Search.

```python
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

Her adımda arama alanını yarıya böldüğü için O(log n) zaman karmaşıklığına sahiptir.

**O(n log n) – Lineer Logaritmik Zaman**

Tanım: Çoğu verimli sıralama algoritmasının zaman karmaşıklığıdır.

Örnek: Merge Sort ve Quick Sort.

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1
    return arr
```

Merge Sort, her bölümleme işlemi O(log n) ve birleştirme işlemi O(n) olduğundan O(n log n) karmaşıklığına sahiptir.

**O(2ⁿ) – Üstel Zaman**

Tanım: Girdi boyutu arttıkça işlem sayısının üstel olarak arttığı algoritmalar. Genellikle rekürsif çözüm yöntemlerinde görülür.

Örnek: Rekürsif yöntemle Fibonacci dizisini hesaplamak.

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
```

Rekürsif olarak çözüm arandığı için zaman karmaşıklığı O(2ⁿ) olur

**O(n!) – Faktöriyel Zaman**

Tanım: Çözüm uzayını tam taramak gereken durumlarda kullanılır.

Örnek: Permütasyon üretme algoritmaları.

```python
def generate_permutations(arr):
    if len(arr) == 0:
        return [[]]
    first = arr[0]
    rest = generate_permutations(arr[1:])
    permutations = []
    for perm in rest:
        for i in range(len(perm) + 1):
            permutations.append(perm[:i] + [first] + perm[i:])
    return permutations
```

Tüm olası sıralamaları (permütasyonları) bulduğu için faktöriyel zaman karmaşıklığına sahiptir.

##**Karşılaştırmalı Big O**

$O(1) < O(log n) < O(n) < O(n log n) < O(n²) < O(2ⁿ) < O(n!)$

##Uygulamalar

**Veri arama:** Binary Search $(O(log n))$

**Veri sıralama:** Merge Sort $(O(n log n))$, Insertion Sort $(O(n²))$

**Optimizasyon:** Dinamik programlama ile üstel zaman $(O(2ⁿ))$ yerine polinom zamanda $(O(n²))$ çözümler bulunabilir.


**Asimptotik Notasyonlar**

Big O Notation, bir algoritmanın performansını 3 farklı durumda analiz eder:


**Best-case:** En iyi senaryo / Omega notasyonu / Omega gösterimi / Big-Ω


**Average-case:** Ortalama senaryo / Teta notasyonu / Teta gösterimi / Big-θ


**Worst-case:** En kötü senaryo / Big-O notasyonu / Büyük O Notasyonu / Big O gösterimi

##Big Omega (Ω) Notasyonu

Anlamı: Bir algoritmanın en iyi durumda (best case) ne kadar hızlı çalışacağının alt sınırını verir. Başka bir deyişle, algoritma asla bu süreden daha kısa sürede çalışamaz.

Matematiksel Tanım: $f(n) = Ω(g(n))$ ise, pozitif sabitler c ve n₀ için, tüm $n ≥ n₀$ değerleri için $0 ≤ c * g(n) ≤ f(n)$ eşitsizliği sağlanır.

Yani algoritmanın çalışma süresi, $g(n)$ fonksiyonuyla orantılı veya daha hızlıdır.

Örnek: Bir sıralı dizide bir eleman arama işlemi için en iyi durum, aranılan elemanın dizinin başında olmasıdır. Bu durumda, sadece bir karşılaştırma yapılacağı için zaman karmaşıklığı $Ω(1)$'dir.

##Big Theta (Θ) Notasyonu

Anlamı: Bir algoritmanın hem en iyi hem de en kötü durumdaki çalışma süresini sıkı bir şekilde sınırlar. Yani, algoritmanın çalışma süresi her zaman belirli bir büyüme oranına sahiptir.

Matematiksel Tanım: $f(n) = Θ(g(n))$ ise, pozitif sabitler $c₁, c₂$ ve $n₀$ için, tüm $n ≥ n₀$ değerleri için $0 ≤ c₁ * g(n) ≤ f(n) ≤ c₂ * g(n)$ eşitsizliği sağlanır.

Ne Anlama Gelir: Algoritmanın çalışma süresi, $g(n)$ fonksiyonuyla aynı büyüme oranına sahiptir.

Örnek: Doğrusal arama algoritması için, en iyi ve en kötü durumlar arasında çok büyük bir fark yoktur. Bu nedenle, doğrusal arama algoritmasının zaman karmaşıklığı $Θ(n)$ olarak ifade edilir.


###Örneklerle Anlatım

Sıralı Dizi İçinde Arama

En iyi durum: Aranan eleman dizinin başındaysa, zaman karmaşıklığı Ω(1) ve O(1) olur. Yani, Θ(1) olarak ifade edilebilir.

En kötü durum: Aranan eleman dizide yoksa veya en sonda ise, zaman karmaşıklığı O(n) olur.

Sıralama Algoritmaları

Merge Sort: En iyi, ortalama ve en kötü durumlar için zaman karmaşıklığı Θ(n log n) olarak ifade edilir.

Bubble Sort: En iyi durum için Ω(n), en kötü durum için O(n²) olarak ifade edilir. Bu nedenle, sıkı bir sınır olmadığı için Θ notasyonu kullanılamaz.
