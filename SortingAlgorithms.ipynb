Insertion Sort, gÃ¼nlÃ¼k hayatta fark etmeden kullandÄ±ÄŸÄ±mÄ±z bir yÃ¶nteme benzetilebilir; Ã¶rneÄŸin, iskambil kartlarÄ±nÄ± sÄ±raya dizerken her kartÄ± uygun yerine yerleÅŸtirme mantÄ±ÄŸÄ± gibi.

**AlgoritmanÄ±n Genel Ã‡alÄ±ÅŸma Prensibi**

Algoritma, dizinin ilk elemanÄ±nÄ± sÄ±ralÄ± kabul eder.
Ä°kinci elemandan baÅŸlayarak, her elemanÄ± dizinin sÄ±ralÄ± kÄ±smÄ±yla karÅŸÄ±laÅŸtÄ±rÄ±r ve uygun konuma yerleÅŸtirir.

Dizinin sonuna kadar devam eden bu sÃ¼reÃ§te, bÃ¼yÃ¼k elemanlar saÄŸa kaydÄ±rÄ±larak kÃ¼Ã§Ã¼k elemanlarÄ±n yerleÅŸmesi saÄŸlanÄ±r.
Insertion Sort, kÃ¼Ã§Ã¼k veri kÃ¼meleri iÃ§in hÄ±zlÄ± ve etkilidir, Ã§Ã¼nkÃ¼ her bir eleman yalnÄ±zca kendi Ã¶nÃ¼ndeki elemanlarla kÄ±yaslanÄ±r.
Ancak, veri kÃ¼mesinin boyutu arttÄ±kÃ§a karÅŸÄ±laÅŸtÄ±rma sayÄ±sÄ± katlanarak artar, bu da algoritmanÄ±n zaman karmaÅŸÄ±klÄ±ÄŸÄ±nÄ±
$ğ‘‚(ğ‘›^2)$ yapar. Bu nedenle bÃ¼yÃ¼k veri kÃ¼meleri iÃ§in daha verimli sÄ±ralama algoritmalarÄ±na ihtiyaÃ§ duyulur.

**Ã–zellikler**

KararlÄ± (Stable): AynÄ± deÄŸere sahip elemanlar sÄ±ralama sonrasÄ± sÄ±ralandÄ±klarÄ± sÄ±rayÄ± korur.

Yerinde SÄ±ralama (In-Place Sort): Ekstra bellek kullanmaz, sÄ±ralamayÄ± mevcut dizi Ã¼zerinde yapar.

En Ä°yi Durum KarmaÅŸÄ±klÄ±ÄŸÄ±: EÄŸer dizi baÅŸtan sona sÄ±ralÄ±ysa, O(n) karmaÅŸÄ±klÄ±ÄŸÄ±yla Ã§alÄ±ÅŸÄ±r.

```python
def insertion_sort(arr):
    # Dizinin uzunluÄŸu kadar dÃ¶ngÃ¼ Ã§alÄ±ÅŸtÄ±r
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        # SÄ±ralanmÄ±ÅŸ diziyi geriye doÄŸru kontrol et
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1

        # DoÄŸru yere yerleÅŸtir
        arr[j + 1] = key
    return arr

# Test
arr = [12, 11, 13, 5, 6]
print("Orijinal dizi:", arr)
insertion_sort(arr)
print("SÄ±ralanmÄ±ÅŸ dizi:", arr)
```

Ã–rnek: ÅÃ¶yle bir dizi dÃ¼ÅŸÃ¼nelim: [38, 27, 43, 3, 9, 82, 10]

BÃ¼yÃ¼k Veri KÃ¼melerinde Verimli: Ã–zellikle Ã§ok bÃ¼yÃ¼k veri kÃ¼melerinde Merge Sort, zaman karmaÅŸÄ±klÄ±ÄŸÄ±ndan dolayÄ± etkili bir ÅŸekilde kullanÄ±labilir.

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        # Sol ve saÄŸ yarÄ±yÄ± sÄ±rala
        merge_sort(left_half)
        merge_sort(right_half)

        # BirleÅŸtir
        i = j = k = 0

        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1
    return arr

# Ã–rnek
arr = [12, 11, 13, 5, 6, 7]
print("Orijinal dizi:", arr)
merge_sort(arr)
print("SÄ±ralanmÄ±ÅŸ dizi:", arr)
```

# **SELECTION SORT**

Selection Sort (SeÃ§meli SÄ±ralama), sÄ±ralanacak dizideki elemanlardan en kÃ¼Ã§Ã¼ÄŸÃ¼nÃ¼ ya da duruma gÃ¶re en bÃ¼yÃ¼ÄŸÃ¼nÃ¼ seÃ§ip, sÄ±ralanmÄ±ÅŸ kÄ±sma ekleyen bir sÄ±ralama algoritmasÄ±dÄ±r. Bu iÅŸlem, dizinin her bir elemanÄ± iÃ§in tekrar edilir. BasitliÄŸi nedeniyle genellikle Ã¶ÄŸrenme amaÃ§lÄ± kullanÄ±lsa da, bÃ¼yÃ¼k veri setlerinde Ã§ok verimli deÄŸildir.

**AdÄ±m AdÄ±m Ã‡alÄ±ÅŸma MantÄ±ÄŸÄ±**

Diziyi Tarama: Algoritma, dizinin sÄ±ralanmamÄ±ÅŸ kÄ±smÄ±ndaki en kÃ¼Ã§Ã¼k (veya bÃ¼yÃ¼k) elemanÄ± bulur.


Yer DeÄŸiÅŸtirme: Bulunan bu eleman, sÄ±ralanmamÄ±ÅŸ kÄ±smÄ±n baÅŸÄ±ndaki elemanla yer deÄŸiÅŸtirir. Bu ÅŸekilde, sÄ±ralanan kÄ±sma bir eleman daha eklenmiÅŸ olur.


AdÄ±mlarÄ± Tekrarlama: Algoritma diziyi baÅŸtan sona tarar, her adÄ±mda sÄ±ralanmamÄ±ÅŸ kÄ±sÄ±mdaki en kÃ¼Ã§Ã¼k elemanÄ± bulup sÄ±ralÄ± kÄ±smÄ±n sonuna ekler. Bu iÅŸlem dizinin tÃ¼m elemanlarÄ± sÄ±ralanana kadar devam eder.


**Ã–rnek:**
Bir dizimiz olsun: [64, 25, 12, 22, 11]


Ä°lk AdÄ±m:

Dizinin sÄ±ralanmamÄ±ÅŸ hali: [64, 25, 12, 22, 11]
En kÃ¼Ã§Ã¼k eleman 11'dir, bu eleman ilk sÄ±radaki elemanla (64) yer deÄŸiÅŸtirir.

SonuÃ§: [11, 25, 12, 22, 64]

Ä°kinci AdÄ±m:

SÄ±ralanmamÄ±ÅŸ kÄ±sÄ±m: [25, 12, 22, 64]
En kÃ¼Ã§Ã¼k eleman 12, ilk sÄ±radaki 25 ile yer deÄŸiÅŸtirir.

ÃœÃ§Ã¼ncÃ¼ AdÄ±m:

En kÃ¼Ã§Ã¼k eleman 22, 25 ile yer deÄŸiÅŸtirir.

DÃ¶rdÃ¼ncÃ¼ AdÄ±m:

**Zaman KarmaÅŸÄ±klÄ±ÄŸÄ±**
En kÃ¶tÃ¼ durumda ve ortalama durumda: O(nÂ²)

En iyi durumda: O(nÂ²) (Yani dizi zaten sÄ±ralÄ± olsa bile algoritma her elemanÄ± kontrol eder.)

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        # SÄ±ralanmamÄ±ÅŸ dizinin en kÃ¼Ã§Ã¼k elemanÄ±nÄ± bul
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j

        # Bulunan en kÃ¼Ã§Ã¼k elemanÄ± sÄ±ralanmamÄ±ÅŸ parÃ§anÄ±n ilk elemanÄ±yla deÄŸiÅŸtir
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

# Ã–rnek
if __name__ == "__main__":
    array = [64, 25, 12, 22, 11]
    print("Orijinal dizi:", array)
    selection_sort(array)
    print("SÄ±ralanmÄ±ÅŸ dizi:", array)
```

**AlgoritmanÄ±n AdÄ±mlarÄ±:**

## **HEAP SORT**

Heap Sort, veri yapÄ±sÄ± olarak heap kullanarak sÄ±ralama iÅŸlemi gerÃ§ekleÅŸtiren bir algoritmadÄ±r. Heap, her dÃ¼ÄŸÃ¼mÃ¼n, Ã§ocuk dÃ¼ÄŸÃ¼mlerinden daha bÃ¼yÃ¼k veya daha kÃ¼Ã§Ã¼k olduÄŸu bir aÄŸaÃ§ yapÄ±sÄ±dÄ±r. **AlgoritmanÄ±n AdÄ±mlarÄ±**

Ä°lk olarak, verilen diziyi bir heap yapÄ±sÄ±na dÃ¶nÃ¼ÅŸtÃ¼rÃ¼rÃ¼z. Bu iÅŸlem, diziyi bir max-heap (veya min-heap) haline getirir.

Sonra kÃ¶k dÃ¼ÄŸÃ¼mÃ¼, heap'in son elemanÄ± ile deÄŸiÅŸtiririz ve heap'in boyutunu bir azaltÄ±rÄ±z.
Heap Ã¶zelliklerini koruyacak ÅŸekilde heap'i yeniden dÃ¼zenleriz buna "heapify" denilir.

3. Tekrarla

Bu iÅŸlemi heap'in boyutu bir eleman kalana kadar tekrar ederiz.

**Teknik Detaylar**

**Heapify**: Verilen bir diziyi heap yapÄ±sÄ±na dÃ¶nÃ¼ÅŸtÃ¼rÃ¼rken, her dÃ¼ÄŸÃ¼mÃ¼ bottom-up ÅŸekilde deÄŸerlendiririz. Bu iÅŸlem, heap'in tÃ¼m Ã¶zelliklerini koruyarak tÃ¼m diziyi tarar.


**Swap:** Heap'ten eleman Ã§Ä±karma iÅŸlemi sÄ±rasÄ±nda, kÃ¶k dÃ¼ÄŸÃ¼m ile dizinin sonundaki elemanÄ± takas ederiz.

**Zaman KarmaÅŸÄ±klÄ±ÄŸÄ±**

Heap Sort, veri yapÄ± olarak binary heap kullanarak sÄ±ralama yapan bir algoritmadÄ±r. Ã–ncelikle veri kÃ¼mesini bir heap yapÄ±sÄ±na dÃ¶nÃ¼ÅŸtÃ¼rÃ¼r, ardÄ±ndan en bÃ¼yÃ¼k (veya en kÃ¼Ã§Ã¼k) Ã¶ÄŸeleri Ã§Ä±kararak sÄ±ralar.

**En KÃ¶tÃ¼ Durum (Worst Case):** O(n logn)

**Ortalama Durum (Average Case):**  O(n logn)

**En Ä°yi Durum (Best Case):** O(n logn)

Heap Sort'un zaman karmaÅŸÄ±klÄ±ÄŸÄ±, veri kÃ¼mesinin dÃ¼zenlenme durumundan baÄŸÄ±msÄ±z olarak O(n logn)'dir. Bu, algoritmanÄ±n tÃ¼m durumlarda aynÄ± performansÄ± gÃ¶sterdiÄŸi anlamÄ±na gelir.

```python
def heapify(arr, n, i):
    largest = i  # KÃ¶kÃ¼ en bÃ¼yÃ¼k olarak kabul et
    left = 2 * i + 1     # Sol Ã§ocuk
    right = 2 * i + 2    # SaÄŸ Ã§ocuk

    # Sol Ã§ocuk bÃ¼yÃ¼kse kÃ¶kten daha bÃ¼yÃ¼k olanÄ± belirle
    if left < n and arr[i] < arr[left]:
        largest = left

    # SaÄŸ Ã§ocuk bÃ¼yÃ¼kse kÃ¶kten daha bÃ¼yÃ¼k olanÄ± belirle
    if right < n and arr[largest] < arr[right]:
        largest = right

    # KÃ¶k en bÃ¼yÃ¼k deÄŸilse, kÃ¶kÃ¼ ve en bÃ¼yÃ¼k Ã§ocuklarÄ± takas et
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]  # Takas
        heapify(arr, n, largest)  # AÅŸaÄŸÄ±da heapify uygula

def heap_sort(arr):
    n = len(arr)

    # Heap oluÅŸturma (bottom-up yaklaÅŸÄ±mÄ±)
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    # ElemanlarÄ± birer birer Ã§Ä±karma ve sÄ±ralÄ± diziyi oluÅŸturma
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]  # Takas
        heapify(arr, i, 0)  # Kalan heap'i yeniden dÃ¼zenle

# Ã–rnek kullanÄ±m
arr = [12, 11, 13, 5, 6, 7]
print("Orijinal dizi:", arr)

heap_sort(arr)
print("SÄ±ralÄ± dizi:", arr)
```

SÄ±ralÄ± run'lar, merge sort kullanÄ±larak birleÅŸtirilir. Bu adÄ±mda, sÄ±ralÄ± run'lar birleÅŸtirilir ve tek bir sÄ±ralÄ± dizi oluÅŸturulur.

**Teknik Detaylar**

Min Run: Timsort, diziyi kÃ¼Ã§Ã¼k run'lara bÃ¶ler ve her run'Ä± sÄ±ralar. Min run uzunluÄŸu genellikle 32 veya 64 olarak seÃ§ilir. Run uzunluÄŸunun Ã§ok kÄ±sa olmasÄ± durumunda, algoritma daha bÃ¼yÃ¼k run'lar oluÅŸturur.


Insertion Sort: KÃ¼Ã§Ã¼k run'lar iÃ§in hÄ±zlÄ± ve verimli bir sÄ±ralama yÃ¶ntemi olarak kullanÄ±lÄ±r.

Merge Sort: BÃ¼yÃ¼k run'larÄ± birleÅŸtirirken, verimli bir ÅŸekilde sÄ±ralÄ± hale getirir.


**Zaman KarmaÅŸÄ±klÄ±ÄŸÄ±**

Timsort, hem Merge Sort hem de Insertion Sort algoritmalarÄ±nÄ± birleÅŸtiren bir sÄ±ralama algoritmasÄ±dÄ±r. Ã–zellikle gerÃ§ek dÃ¼nyadaki veri kÃ¼melerinde yÃ¼ksek verimlilik gÃ¶sterir.

**En KÃ¶tÃ¼ Durum (Worst Case):** O(n logn)

**Ortalama Durum (Average Case):**  O(n logn)

**En Ä°yi Durum (Best Case):** O(n logn)

Ancak, Timsort genellikle sÄ±ralÄ± alt dizilerde daha hÄ±zlÄ± Ã§alÄ±ÅŸÄ±r, bu nedenle en iyi durum performansÄ± kÄ±smi sÄ±ralama iÃ§in daha hÄ±zlÄ± olabilir.

Timsort'un gerÃ§ek dÃ¼nya verileriyle daha hÄ±zlÄ± ve daha verimli Ã§alÄ±ÅŸtÄ±ÄŸÄ± yaygÄ±n olarak kabul edilir.

```python
def insertion_sort(arr, left, right):
    for i in range(left + 1, right + 1):
        key = arr[i]
        j = i - 1
        while j >= left and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

def merge(arr, l, m, r):
    len1, len2 = m - l + 1, r - m
    left, right = arr[l:l + len1], arr[m + 1:m + 1 + len2]

    i, j, k = 0, 0, l
    while i < len1 and j < len2:
        if left[i] <= right[j]:
            arr[k] = left[i]
            i += 1
        else:
            arr[k] = right[j]
            j += 1
        k += 1

    while i < len1:
        arr[k] = left[i]
        i += 1
        k += 1

    while j < len2:
        arr[k] = right[j]
        j += 1
        k += 1

def tim_sort(arr):
    min_run = 32
    n = len(arr)

    # 1. KÃ¼Ã§Ã¼k parÃ§alara ayÄ±rma ve sÄ±ralama
    for i in range(0, n, min_run):
        insertion_sort(arr, i, min((i + min_run - 1), n - 1))

    # 2. Run'larÄ± birleÅŸtirme
    size = min_run
    while size < n:
        for start in range(0, n, 2 * size):
            mid = min(start + size - 1, n - 1)
            end = min(start + 2 * size - 1, n - 1)
            if mid < end:
                merge(arr, start, mid, end)
        size *= 2

# Ã–rnek kullanÄ±m
arr = [5, 21, 7, 23, 19, 75, 13, 9, 82, 36, 78, 108, 93, 1, 25, 14]
print("Orijinal dizi:", arr)
tim_sort(arr)
print("SÄ±ralÄ± dizi:", arr)
```
